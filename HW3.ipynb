```python
from google.colab import userdata
import os
import uuid
import json
import pandas as pd
from datetime import datetime as dt
from dateutil.tz import gettz
import gradio as gr

# ========================
# GEMINI KEY
# ========================
gemini_key = userdata.get('gemini')
os.environ["GEMINI_API_KEY"] = gemini_key

# ========================
# Google Sheet ç›¸é—œ
# ========================
def ensure_worksheet(sh, title, header):
    try:
        ws = sh.worksheet(title)
    except gspread.WorksheetNotFound:
        ws = sh.add_worksheet(title=title, rows="1000", cols=str(len(header)+5))
        ws.update([header])
    data = ws.get_all_values()
    if not data or (data and data[0] != header):
        ws.clear()
        ws.update([header])
    return ws

TASKS_HEADER = [
    "id","task","status","priority","est_min","start_time","end_time",
    "actual_min","pomodoros","due_date","labels","notes",
    "created_at","updated_at","completed_at","planned_for"
]
LOGS_HEADER = [
    "log_id","task_id","phase","start_ts","end_ts","minutes","cycles","note"
]

ws_tasks = ensure_worksheet(sh, "tasks", TASKS_HEADER)
ws_logs  = ensure_worksheet(sh, "pomodoro_logs", LOGS_HEADER)

def tznow():
    return dt.now(gettz(TIMEZONE))

def read_df(ws, header):
    df = get_as_dataframe(ws, evaluate_formulas=True, header=0)
    if df is None or df.empty:
        return pd.DataFrame(columns=header)
    df = df.fillna("")
    for c in header:
        if c not in df.columns:
            df[c] = ""
    if "est_min" in df.columns:
        df["est_min"] = pd.to_numeric(df["est_min"], errors="coerce").fillna(0).astype(int)
    if "actual_min" in df.columns:
        df["actual_min"] = pd.to_numeric(df["actual_min"], errors="coerce").fillna(0).astype(int)
    if "pomodoros" in df.columns:
        df["pomodoros"] = pd.to_numeric(df["pomodoros"], errors="coerce").fillna(0).astype(int)
    return df[header]

def write_df(ws, df, header):
    if df.empty:
        ws.clear()
        ws.update([header])
        return
    df_out = df.copy()
    for c in df_out.columns:
        df_out[c] = df_out[c].astype(str)
    ws.clear()
    ws.update([header] + df_out.values.tolist())

def refresh_all():
    return (
        read_df(ws_tasks, TASKS_HEADER).copy(),
        read_df(ws_logs, LOGS_HEADER).copy(),
        None  # ç§»é™¤ Clips
    )

tasks_df, logs_df, _ = refresh_all()

# ========================
# Tasks æ“ä½œ
# ========================
def add_task(task, priority, est_min, due_date, labels, notes, planned_for):
    global tasks_df
    _now = tznow().isoformat()
    new = pd.DataFrame([{
        "id": str(uuid.uuid4())[:8],
        "task": task.strip(),
        "status": "todo",
        "priority": priority or "M",
        "est_min": int(est_min) if est_min else 25,
        "start_time": "",
        "end_time": "",
        "actual_min": 0,
        "pomodoros": 0,
        "due_date": due_date or "",
        "labels": labels or "",
        "notes": notes or "",
        "created_at": _now,
        "updated_at": _now,
        "completed_at": "",
        "planned_for": planned_for or ""
    }])
    tasks_df = pd.concat([tasks_df, new], ignore_index=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… å·²æ–°å¢ä»»å‹™", tasks_df

def update_task_status(task_id, new_status):
    global tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx)==0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    i = idx[0]
    tasks_df.loc[i, "status"] = new_status
    tasks_df.loc[i, "updated_at"] = tznow().isoformat()
    if new_status == "done" and not tasks_df.loc[i, "completed_at"]:
        tasks_df.loc[i, "completed_at"] = tznow().isoformat()
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… ç‹€æ…‹å·²æ›´æ–°", tasks_df

def mark_done(task_id):
    return update_task_status(task_id, "done")

def recalc_task_actuals(task_id):
    global tasks_df, logs_df
    work_logs = logs_df[(logs_df["task_id"]==task_id) & (logs_df["phase"]=="work")]
    total_min = work_logs["minutes"].astype(float).sum() if not work_logs.empty else 0
    pomos = int(round(total_min / 25.0))
    idx = tasks_df.index[tasks_df["id"]==task_id]
    if len(idx)==0: return
    i = idx[0]
    tasks_df.loc[i,"actual_min"] = int(total_min)
    tasks_df.loc[i,"pomodoros"] = pomos
    tasks_df.loc[i,"updated_at"] = tznow().isoformat()

def list_task_choices():
    global tasks_df
    if tasks_df.empty:
        return []
    def row_label(r):
        return f"[{r['status']}] (P:{r['priority']}) {r['task']} â€” {r['id']}"
    return [(row_label(r), r["id"]) for _, r in tasks_df.iterrows()]

_active_sessions = {}

def start_phase(task_id, phase, cycles):
    if not task_id: return "âš ï¸ è«‹å…ˆé¸æ“‡ä»»å‹™"
    _active_sessions[task_id] = {
        "phase": phase,
        "start_ts": tznow().isoformat(),
        "cycles": int(cycles) if cycles else 1
    }
    return f"â–¶ï¸ å·²é–‹å§‹ï¼š{phase}ï¼ˆtask: {task_id}ï¼‰"

def end_phase(task_id, note):
    global logs_df, tasks_df
    if task_id not in _active_sessions:
        return "âš ï¸ å°šæœªé–‹å§‹ä»»ä½•éšæ®µ"
    sess = _active_sessions.pop(task_id)
    start = pd.to_datetime(sess["start_ts"])
    end = tznow()
    minutes = round((end - start).total_seconds() / 60.0, 2)
    log = pd.DataFrame([{
        "log_id": str(uuid.uuid4())[:8],
        "task_id": task_id,
        "phase": sess["phase"],
        "start_ts": start.isoformat(),
        "end_ts": end.isoformat(),
        "minutes": minutes,
        "cycles": int(sess["cycles"]),
        "note": note or ""
    }])
    logs_df = pd.concat([logs_df, log], ignore_index=True)
    write_df(ws_logs, logs_df, LOGS_HEADER)
    if sess["phase"] == "work":
        recalc_task_actuals(task_id)
        write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return f"â¹ï¸ å·²çµæŸï¼š{sess['phase']}ï¼Œç´€éŒ„ {minutes} åˆ†é˜"

def generate_today_plan():
    global tasks_df
    today = tznow().date().isoformat()
    cand = tasks_df[((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today")) & (tasks_df["status"]!="done")].copy()
    if cand.empty:
        return "ğŸ“­ ä»Šå¤©æ²’æœ‰æ¨™è¨˜çš„ä»»å‹™ã€‚è«‹åœ¨ Tasks åˆ†é æŠŠä»»å‹™çš„ due_date è¨­ç‚ºä»Šå¤©æˆ– planned_for è¨­ç‚º todayã€‚"

    pr_order = {"H":0, "M":1, "L":2}
    cand["p_ord"] = cand["priority"].map(pr_order).fillna(3)
    cand = cand.sort_values(["p_ord","est_min"], ascending=[True, True])

    plan_md = "ğŸ”§ æœªè¨­å®š GEMINI_API_KEYï¼Œä½¿ç”¨è¦å‰‡å¼è¦åŠƒã€‚\n\n"
    api_key = os.environ.get("GEMINI_API_KEY","").strip()
    if api_key:
        try:
            genai.configure(api_key=api_key)
            sys_prompt = (
                "ä½ æ˜¯ä¸€ä½ä»»å‹™è¦åŠƒåŠ©ç†ã€‚è«‹æŠŠè¼¸å…¥çš„ä»»å‹™ï¼ˆå«ä¼°æ™‚èˆ‡å„ªå…ˆç´šï¼‰æ’æˆä¸‰æ®µï¼šmorningã€afternoonã€eveningï¼Œ"
                "ä¸¦çµ¦å‡ºæ¯æ®µçš„é‡é»ã€é †åºã€æ¯é …çš„æ™‚é–“é ä¼°èˆ‡å‚™è¨»ã€‚ç¸½æ™‚æ•¸è«‹å¤§è‡´ç¬¦åˆä»»å‹™ä¼°æ™‚ç¸½å’Œã€‚"
                "å›å‚³ä»¥ Markdown æ¢åˆ—ï¼Œæ ¼å¼ï¼š\n"
                "### Morning\n- [ä»»å‹™ID] ä»»å‹™åç¨±ï¼ˆé ä¼° xx åˆ†ï¼‰â€” å‚™è¨»\n... "
            )
            items = [{"id": r["id"], "task": r["task"], "est_min": int(r["est_min"]), "priority": r["priority"]} for _, r in cand.iterrows()]
            user_content = json.dumps({"today": today, "tasks": items}, ensure_ascii=False)
            model = genai.GenerativeModel("gemini-2.0-flash")
            resp = model.generate_content(sys_prompt + "\n\n" + user_content)
            plan_md = resp.text
        except Exception as e:
            plan_md = f"âš ï¸ Gemini å¤±æ•—ï¼š{e}\n\næ”¹ç”¨è¦å‰‡å¼è¦åŠƒã€‚"

    buckets = {"morning": [], "afternoon": [], "evening": []}
    for i, (_, r) in enumerate(cand.iterrows()):
        if i % 3 == 0: buckets["morning"].append(r)
        elif i % 3 == 1: buckets["afternoon"].append(r)
        else: buckets["evening"].append(r)

    def sec_md(name, rows):
        if not rows: return f"### {name.title()}\nï¼ˆç„¡ï¼‰\n"
        lines = [f"### {name.title()}"]
        for r in rows:
            lines.append(f"- [{r['id']}] {r['task']}ï¼ˆé ä¼° {int(r['est_min'])} åˆ†ï¼ŒP:{r['priority']}ï¼‰")
        return "\n".join(lines) + "\n"

    rule_md = sec_md("morning", buckets["morning"]) + "\n" + sec_md("afternoon", buckets["afternoon"]) + "\n" + sec_md("evening", buckets["evening"])
    return (plan_md + "\n---\n" + rule_md).strip()

def today_summary():
    global tasks_df
    today = tznow().date().isoformat()
    planned = tasks_df[((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today"))]
    done = planned[planned["status"]=="done"]
    total = len(planned)
    done_n = len(done)
    rate = (done_n/total*100) if total>0 else 0
    return f"ğŸ“… ä»Šæ—¥è¨ˆç•«ä»»å‹™ï¼š{total}ï¼›âœ… å®Œæˆï¼š{done_n}ï¼›ğŸ“ˆ å®Œæˆç‡ï¼š{rate:.1f}%"

def delete_task(task_id):
    global tasks_df
    if not task_id:
        return "âš ï¸ è«‹å…ˆé¸æ“‡è¦åˆªé™¤çš„ä»»å‹™", tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx) == 0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    tasks_df = tasks_df.drop(idx).reset_index(drop=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return f"ğŸ—‘ï¸ å·²åˆªé™¤ä»»å‹™ï¼š{task_id}", tasks_df

def query_tasks(keyword, status_filter, priority_filter, start_due, end_due):
    global tasks_df
    df = tasks_df.copy()
    if status_filter:
        df = df[df["status"].fillna("").str.lower().isin([s.lower() for s in status_filter])]
    if priority_filter and priority_filter.strip():
        df = df[df["priority"].fillna("").str.upper() == priority_filter.upper()]
    if keyword and keyword.strip():
        kw = keyword.lower()
        df = df[df["task"].str.lower().str.contains(kw, na=False) | df["notes"].str.lower().str.contains(kw, na=False)]
    df_dates = pd.to_datetime(df["due_date"], errors="coerce")
    if start_due:
        start_dt = pd.to_datetime(start_due)
        df = df[df_dates >= start_dt]
    if end_due:
        end_dt = pd.to_datetime(end_due)
        df = df[df_dates <= end_dt]
    if df.empty:
        return "ğŸ“­ æ²’æœ‰ç¬¦åˆæ¢ä»¶çš„ä»»å‹™", pd.DataFrame(columns=tasks_df.columns)
    df = df.sort_values("due_date", ascending=True)
    return f"âœ… æ‰¾åˆ° {len(df)} ç­†ä»»å‹™", df

def export_tasks_logs(format="csv"):
    global tasks_df, logs_df
    _now = dt.now().strftime("%Y%m%d_%H%M%S")
    if format=="csv":
        tasks_file = f"tasks_{_now}.csv"
        logs_file  = f"logs_{_now}.csv"
        tasks_df.to_csv(tasks_file, index=False)
        logs_df.to_csv(logs_file, index=False)
    else:
        tasks_file = f"tasks_{_now}.json"
        logs_file  = f"logs_{_now}.json"
        tasks_df.to_json(tasks_file, orient="records", force_ascii=False, indent=2)
        logs_df.to_json(logs_file, orient="records", force_ascii=False, indent=2)
    return tasks_file, logs_file

def import_tasks_logs(tasks_file, logs_file, format="csv"):
    global tasks_df, logs_df, ws_tasks, ws_logs
    try:
        if tasks_file is None and logs_file is None:
            return "âš ï¸ è«‹è‡³å°‘ä¸Šå‚³ä¸€å€‹æª”æ¡ˆï¼ˆä»»å‹™æˆ–ç´€éŒ„ï¼‰"
        if tasks_file:
            if format=="csv": df_tasks = pd.read_csv(tasks_file.name)
            else:
                with open(tasks_file.name, "r", encoding="utf-8") as f:
                    df_tasks = pd.DataFrame(json.load(f))
            write_df(ws_tasks, df_tasks, TASKS_HEADER)
            tasks_df = df_tasks
        if logs_file:
            if format=="csv": df_logs = pd.read_csv(logs_file.name)
            else:
                with open(logs_file.name, "r", encoding="utf-8") as f:
                    df_logs = pd.DataFrame(json.load(f))
            write_df(ws_logs, df_logs, LOGS_HEADER)
            logs_df = df_logs
        return "âœ… åŒ¯å…¥å®Œæˆï¼Œè³‡æ–™å·²å¯«å…¥ Google Sheet ä¸¦åŒæ­¥åˆ°ç•«é¢"
    except Exception as e:
        return f"âš ï¸ åŒ¯å…¥å¤±æ•—ï¼š{e}"

def export_query_results(df):
    if df is None or len(df)==0: return None
    _now = dt.now().strftime("%Y%m%d_%H%M%S")
    filename = f"query_results_{_now}.csv"
    df.to_csv(filename, index=False)
    return filename

def _refresh():
    global tasks_df, logs_df
    tasks_df, logs_df, _ = refresh_all()
    choices = list_task_choices()
    summary = today_summary()
    return tasks_df, logs_df, gr.update(choices=choices), gr.update(choices=choices), summary

# ========================
# Gradio ä»‹é¢
# ========================

with gr.Blocks(title="å¾…è¾¦æ¸…å–®ï¼‹ç•ªèŒ„é˜ï¼‹AI è¨ˆç•«") as demo:
    gr.Markdown("# âœ… å¾…è¾¦æ¸…å–®èˆ‡ç•ªèŒ„é˜ï¼ˆGoogle Sheetï¼‹Gradioï¼‹AI è¨ˆç•«ï¼‰")
    with gr.Row():
        btn_refresh = gr.Button("ğŸ”„ é‡æ–°æ•´ç†ï¼ˆSheet â†’ Appï¼‰")
        out_summary = gr.Markdown(today_summary())

    with gr.Tab("Tasks"):
        with gr.Row():
            with gr.Column(scale=2):
                task = gr.Textbox(label="ä»»å‹™åç¨±")
                priority = gr.Dropdown(["H","M","L"], value="M", label="å„ªå…ˆç´š")
                est_min = gr.Number(value=25, label="é ä¼°æ™‚é–“ï¼ˆåˆ†é˜ï¼‰", precision=0)
                due_date = gr.Textbox(label="åˆ°æœŸæ—¥ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰")
                labels = gr.Textbox(label="æ¨™ç±¤ï¼ˆé€—è™Ÿåˆ†éš”ï¼Œå¯ç©ºç™½ï¼‰")
                notes = gr.Textbox(label="å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
                planned_for = gr.Dropdown(["","today","tomorrow"], value="", label="è¦åŠƒæ­¸å±¬")
                btn_add = gr.Button("â• æ–°å¢ä»»å‹™")
                msg_add = gr.Markdown()
            with gr.Column(scale=3):
                grid_tasks = gr.Dataframe(value=tasks_df, label="ä»»å‹™æ¸…å–®", interactive=False)

        with gr.Row():
            task_choice = gr.Dropdown(choices=list_task_choices(), label="é¸å–ä»»å‹™ï¼ˆç”¨æ–¼æ›´æ–°ï¼‰")
            new_status = gr.Dropdown(["todo","in-progress","done"], value="in-progress", label="æ›´æ–°ç‹€æ…‹")

        with gr.Row():
            btn_update = gr.Button("âœï¸ æ›´æ–°ç‹€æ…‹")
            btn_done = gr.Button("âœ… ç›´æ¥æ¨™è¨˜å®Œæˆ")
            btn_delete = gr.Button("ğŸ—‘ï¸ åˆªé™¤ä»»å‹™")

        msg_update = gr.Markdown()
        msg_delete = gr.Markdown()

    with gr.Tab("Pomodoro"):
        with gr.Row():
            sel_task = gr.Dropdown(choices=list_task_choices(), label="é¸æ“‡ä»»å‹™")
            cycles = gr.Number(value=1, precision=0, label="ç•ªèŒ„æ•¸ï¼ˆåƒ…ä½œç´€éŒ„ï¼‰")
        with gr.Row():
            btn_start_work = gr.Button("â–¶ï¸ é–‹å§‹å·¥ä½œ")
            note_work = gr.Textbox(label="å·¥ä½œå‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_work = gr.Button("â¹ï¸ çµæŸå·¥ä½œä¸¦è¨˜éŒ„")
        with gr.Row():
            btn_start_break = gr.Button("ğŸµ é–‹å§‹ä¼‘æ¯")
            note_break = gr.Textbox(label="ä¼‘æ¯å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_break = gr.Button("â¹ï¸ çµæŸä¼‘æ¯ä¸¦è¨˜éŒ„")
        msg_pomo = gr.Markdown()
        grid_logs = gr.Dataframe(value=logs_df, label="ç•ªèŒ„é˜ç´€éŒ„", interactive=False)

    with gr.Tab("AI Plan"):
        gr.Markdown("æŠŠ**ä»Šå¤©çš„ä»»å‹™**æ’æˆ **morning / afternoon / evening** ä¸‰æ®µè¡Œå‹•è¨ˆç•«ã€‚è‹¥æœªè¨­ GEMINI_API_KEYï¼Œæœƒç”¨è¦å‰‡å¼ã€‚")
        btn_plan = gr.Button("ğŸ§  ç”¢ç”Ÿä»Šæ—¥è¨ˆç•«")
        out_plan = gr.Markdown()

    with gr.Tab("Summary"):
        btn_summary = gr.Button("ğŸ“Š é‡æ–°è¨ˆç®—ä»Šæ—¥å®Œæˆç‡")
        out_summary2 = gr.Markdown()

    with gr.Tab("åŒ¯å…¥ / åŒ¯å‡º"):
        gr.Markdown("ğŸ“‚ åŒ¯å…¥æˆ–åŒ¯å‡ºä»»å‹™èˆ‡ç•ªèŒ„ç´€éŒ„ (CSV / JSON)")

        # ===== åŒ¯å‡ºå€å¡Š =====
        with gr.Row():
            export_format = gr.Radio(["csv", "json"], value="csv", label="åŒ¯å‡ºæ ¼å¼")
            btn_export = gr.Button("ğŸ“¤ åŒ¯å‡º")
        with gr.Row():
            tasks_out = gr.File(label="ä¸‹è¼‰ä»»å‹™æª”æ¡ˆ")
            logs_out = gr.File(label="ä¸‹è¼‰ç´€éŒ„æª”æ¡ˆ")

        # ===== åŒ¯å…¥å€å¡Š =====
        with gr.Row():
            tasks_file = gr.File(label="é¸æ“‡ä»»å‹™æª”æ¡ˆ")
            logs_file  = gr.File(label="é¸æ“‡ç´€éŒ„æª”æ¡ˆ")
            import_format = gr.Radio(["csv","json"], value="csv", label="åŒ¯å…¥æ ¼å¼")
            btn_import = gr.Button("ğŸ“¥ åŒ¯å…¥")
        out_import = gr.Markdown()

    with gr.Tab("æŸ¥è©¢ä»»å‹™"):
        gr.Markdown("ğŸ” **ä»»å‹™æŸ¥è©¢** â€” å¯ä¾ç‹€æ…‹ã€å„ªå…ˆç´šã€é—œéµå­—èˆ‡åˆ°æœŸæ—¥ç¯„åœé€²è¡Œç¯©é¸")

        with gr.Row():
            keyword = gr.Textbox(label="é—œéµå­—æœå°‹ï¼ˆä»»å‹™åç¨±æˆ–å‚™è¨»ï¼‰", placeholder="ä¾‹å¦‚ï¼šå ±å‘Šã€meetingã€å®¢æˆ¶â€¦")

        with gr.Row():
            status_filter = gr.CheckboxGroup(
                ["todo", "in-progress", "done"],
                value=[],
                label="é¸æ“‡ä»»å‹™ç‹€æ…‹ï¼ˆå¯è¤‡é¸ï¼Œå¯ç•™ç©ºï¼‰"
            )
            priority_filter = gr.Dropdown(
                ["", "H", "M", "L"],
                value="",
                label="å„ªå…ˆç´šï¼ˆå¯ç•™ç©ºï¼‰"
            )

        with gr.Row():
            start_due = gr.Textbox(label="åˆ°æœŸæ—¥é–‹å§‹ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰", placeholder="2025-10-01")
            end_due = gr.Textbox(label="åˆ°æœŸæ—¥çµæŸï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰", placeholder="2025-10-31")

        with gr.Row():
            btn_query = gr.Button("ğŸ” æŸ¥è©¢")
            btn_export_query = gr.Button("ğŸ“¤ åŒ¯å‡ºçµæœ (CSV)")

        msg_query = gr.Markdown()
        grid_query = gr.Dataframe(label="æŸ¥è©¢çµæœ", interactive=False)
        file_out = gr.File(label="ä¸‹è¼‰æŸ¥è©¢çµæœ")

    # ç¶å®šäº‹ä»¶
    btn_export.click(export_tasks_logs, inputs=[export_format], outputs=[tasks_out, logs_out])

    btn_import.click(
        import_tasks_logs,
        inputs=[tasks_file, logs_file, import_format],
        outputs=[out_import]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )

    # === ç¶å®šå‹•ä½œ ===
    btn_refresh.click(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )

    btn_add.click(
        add_task,
        inputs=[task, priority, est_min, due_date, labels, notes, planned_for],
        outputs=[msg_add, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )

    btn_delete.click(
        delete_task,
        inputs=[task_choice],
        outputs=[msg_delete, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )

    btn_update.click(
        update_task_status,
        inputs=[task_choice, new_status],
        outputs=[msg_update, grid_tasks]
    )

    btn_done.click(
        mark_done,
        inputs=[task_choice],
        outputs=[msg_update, grid_tasks]
    )

    btn_start_work.click(
        start_phase, inputs=[sel_task, gr.State("work"), cycles], outputs=[msg_pomo]
    )
    btn_end_work.click(
        end_phase, inputs=[sel_task, note_work], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )
    btn_start_break.click(
        start_phase, inputs=[sel_task, gr.State("break"), cycles], outputs=[msg_pomo]
    )
    btn_end_break.click(
        end_phase, inputs=[sel_task, note_break], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, task_choice, sel_task, out_summary]
    )

    btn_plan.click(generate_today_plan, outputs=[out_plan])

    def _crawl_and_save(u, s, m, l):
        df, msg = crawl(u, s, m, l)
        # å¯«å…¥ web_clipsï¼ˆè¦†è“‹å¼è¿½åŠ ï¼šåˆä½µèˆŠè³‡æ–™ï¼‰
        global clips_df
        if not df.empty:
            clips_df = pd.concat([clips_df, df], ignore_index=True)
            write_df(ws_clips, clips_df, CLIPS_HEADER)
        return msg, clips_df

    def _add_clips(clip_ids_str, pr, est):
        ids = [c.strip() for c in (clip_ids_str or "").split(",") if c.strip()]
        msg, new_clips, new_tasks = add_clips_as_tasks(ids, pr, est)
        return msg, new_clips, new_tasks

    btn_summary.click(today_summary, outputs=[out_summary2])

    btn_query.click(
        query_tasks,
        inputs=[keyword, status_filter, priority_filter, start_due, end_due],
        outputs=[msg_query, grid_query]
    )

    btn_export_query.click(
        export_query_results,
        inputs=[grid_query],
        outputs=[file_out]
    )

demo.launch()
```

```python
def ensure_worksheet(sh, title, header):
    try:
        ws = sh.worksheet(title)
    except gspread.WorksheetNotFound:
        ws = sh.add_worksheet(title=title, rows="1000", cols=str(len(header)+5))
        ws.update([header])
    # è‹¥æ²’æœ‰è¡¨é ­å°±è£œä¸Š
    data = ws.get_all_values()
    if not data or (data and data[0] != header):
        ws.clear()
        ws.update([header])
    return ws

TASKS_HEADER = [
    "id","task","status","priority","est_min","start_time","end_time",
    "actual_min","pomodoros","due_date","labels","notes",
    "created_at","updated_at","completed_at","planned_for"
]
LOGS_HEADER = [
    "log_id","task_id","phase","start_ts","end_ts","minutes","cycles","note"
]
CLIPS_HEADER = ["clip_id","url","selector","text","href","created_at","added_to_task"]

ws_tasks = ensure_worksheet(sh, "tasks", TASKS_HEADER)
ws_logs  = ensure_worksheet(sh, "pomodoro_logs", LOGS_HEADER)
ws_clips = ensure_worksheet(sh, "web_clips", CLIPS_HEADER)

def tznow():
    return dt.now(gettz(TIMEZONE))

def read_df(ws, header):
    df = get_as_dataframe(ws, evaluate_formulas=True, header=0)
    if df is None or df.empty:
        return pd.DataFrame(columns=header)
    df = df.fillna("")
    # ä¿è­‰æ¬„ä½é½Šå…¨
    for c in header:
        if c not in df.columns:
            df[c] = ""
    # å‹åˆ¥å¾®èª¿
    if "est_min" in df.columns:
        df["est_min"] = pd.to_numeric(df["est_min"], errors="coerce").fillna(0).astype(int)
    if "actual_min" in df.columns:
        df["actual_min"] = pd.to_numeric(df["actual_min"], errors="coerce").fillna(0).astype(int)
    if "pomodoros" in df.columns:
        df["pomodoros"] = pd.to_numeric(df["pomodoros"], errors="coerce").fillna(0).astype(int)
    return df[header]

def write_df(ws, df, header):
    if df.empty:
        ws.clear()
        ws.update([header])
        return
    # è½‰å­—ä¸²é¿å… gspread å‹åˆ¥å•é¡Œ
    df_out = df.copy()
    for c in df_out.columns:
        df_out[c] = df_out[c].astype(str)
    ws.clear()
    ws.update([header] + df_out.values.tolist())

def refresh_all():
    return (
        read_df(ws_tasks, TASKS_HEADER).copy(),
        read_df(ws_logs, LOGS_HEADER).copy(),
        read_df(ws_clips, CLIPS_HEADER).copy()
    )

tasks_df, logs_df, clips_df = refresh_all()

def add_task(task, priority, est_min, due_date, labels, notes, planned_for):
    global tasks_df
    _now = tznow().isoformat()
    new = pd.DataFrame([{
        "id": str(uuid.uuid4())[:8],
        "task": task.strip(),
        "status": "todo",
        "priority": priority or "M",
        "est_min": int(est_min) if est_min else 25,
        "start_time": "",
        "end_time": "",
        "actual_min": 0,
        "pomodoros": 0,
        "due_date": due_date or "",
        "labels": labels or "",
        "notes": notes or "",
        "created_at": _now,
        "updated_at": _now,
        "completed_at": "",
        "planned_for": planned_for or ""  # å¯å¡« today / tomorrow / ç©ºç™½
    }])
    tasks_df = pd.concat([tasks_df, new], ignore_index=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… å·²æ–°å¢ä»»å‹™", tasks_df

def update_task_status(task_id, new_status):
    global tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx)==0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    i = idx[0]
    tasks_df.loc[i, "status"] = new_status
    tasks_df.loc[i, "updated_at"] = tznow().isoformat()
    if new_status == "done" and not tasks_df.loc[i, "completed_at"]:
        tasks_df.loc[i, "completed_at"] = tznow().isoformat()
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… ç‹€æ…‹å·²æ›´æ–°", tasks_df

def mark_done(task_id):
    return update_task_status(task_id, "done")

def recalc_task_actuals(task_id):
    """æ ¹æ“š logs_df å›å¯« actual_min èˆ‡ pomodoros"""
    global tasks_df, logs_df
    work_logs = logs_df[(logs_df["task_id"]==task_id) & (logs_df["phase"]=="work")]
    total_min = work_logs["minutes"].astype(float).sum() if not work_logs.empty else 0
    pomos = int(round(total_min / 25.0))
    idx = tasks_df.index[tasks_df["id"]==task_id]
    if len(idx)==0: return
    i = idx[0]
    tasks_df.loc[i,"actual_min"] = int(total_min)
    tasks_df.loc[i,"pomodoros"] = pomos
    tasks_df.loc[i,"updated_at"] = tznow().isoformat()

def list_task_choices():
    global tasks_df
    if tasks_df.empty:
        return []
    # é¡¯ç¤ºï¼š [status] (P:priority) task  â€” id
    def row_label(r):
        return f"[{r['status']}] (P:{r['priority']}) {r['task']} â€” {r['id']}"
    return [(row_label(r), r["id"]) for _, r in tasks_df.iterrows()]

# æˆ‘å€‘æ¡ã€ŒæŒ‰éˆ•é–‹å§‹ / çµæŸã€æ¨¡å¼ï¼ˆé¿å…å¾Œç«¯é˜»å¡ï¼‰ï¼Œæ¯æ¬¡æŒ‰ã€Œé–‹å§‹ã€æœƒå…ˆè¨˜ä½ start_tsï¼Œ
# æŒ‰ã€ŒçµæŸã€æ™‚è¨ˆç®—åˆ†é˜ä¸¦å¯«å…¥ logsï¼Œå†å›å¡«ä»»å‹™ actual_min / pomodorosã€‚

_active_sessions = {}  # { task_id: {"phase": "work"/"break", "start_ts": iso, "cycles": int} }

def start_phase(task_id, phase, cycles):
    if not task_id: return "âš ï¸ è«‹å…ˆé¸æ“‡ä»»å‹™"
    _active_sessions[task_id] = {
        "phase": phase,
        "start_ts": tznow().isoformat(),
        "cycles": int(cycles) if cycles else 1
    }
    return f"â–¶ï¸ å·²é–‹å§‹ï¼š{phase}ï¼ˆtask: {task_id}ï¼‰"

def end_phase(task_id, note):
    global logs_df, tasks_df
    if task_id not in _active_sessions:
        return "âš ï¸ å°šæœªé–‹å§‹ä»»ä½•éšæ®µ"
    sess = _active_sessions.pop(task_id)
    start = pd.to_datetime(sess["start_ts"])
    end = tznow()
    minutes = round((end - start).total_seconds() / 60.0, 2)
    log = pd.DataFrame([{
        "log_id": str(uuid.uuid4())[:8],
        "task_id": task_id,
        "phase": sess["phase"],
        "start_ts": start.isoformat(),
        "end_ts": end.isoformat(),
        "minutes": minutes,
        "cycles": int(sess["cycles"]),
        "note": note or ""
    }])
    logs_df = pd.concat([logs_df, log], ignore_index=True)
    write_df(ws_logs, logs_df, LOGS_HEADER)

    # å›å¡«ä»»å‹™
    if sess["phase"] == "work":
        recalc_task_actuals(task_id)
        write_df(ws_tasks, tasks_df, TASKS_HEADER)

    return f"â¹ï¸ å·²çµæŸï¼š{sess['phase']}ï¼Œç´€éŒ„ {minutes} åˆ†é˜"

# AI è¨ˆç•«ï¼ˆGeminiï¼›ç„¡é‡‘é‘°å‰‡è¦å‰‡å¼ï¼‰
def generate_today_plan():
    global tasks_df
    # ä»¥ã€Œdue_date æ˜¯ä»Šå¤©ã€æˆ–ã€Œplanned_for = todayã€ä¸”ä¸æ˜¯ done çš„ä»»å‹™ç‚ºè¨ˆç•«æ¸…å–®
    today = tznow().date().isoformat()
    cand = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today")) &
        (tasks_df["status"]!="done")
    ].copy()
    if cand.empty:
        return "ğŸ“­ ä»Šå¤©æ²’æœ‰æ¨™è¨˜çš„ä»»å‹™ã€‚è«‹åœ¨ Tasks åˆ†é æŠŠä»»å‹™çš„ due_date è¨­ç‚ºä»Šå¤©æˆ– planned_for è¨­ç‚º todayã€‚"

    # å…ˆä¾ priorityï¼ˆH>M>Lï¼‰+ est_min æ’åº
    pr_order = {"H":0, "M":1, "L":2}
    cand["p_ord"] = cand["priority"].map(pr_order).fillna(3)
    cand = cand.sort_values(["p_ord","est_min"], ascending=[True, True])

    # å˜—è©¦ Gemini
    api_key = os.environ.get("GEMINI_API_KEY","").strip()
    if api_key:
        genai.configure(api_key=api_key)
        sys_prompt = (
            "ä½ æ˜¯ä¸€ä½ä»»å‹™è¦åŠƒåŠ©ç†ã€‚è«‹æŠŠè¼¸å…¥çš„ä»»å‹™ï¼ˆå«ä¼°æ™‚èˆ‡å„ªå…ˆç´šï¼‰æ’æˆä¸‰æ®µï¼šmorningã€afternoonã€eveningï¼Œ"
            "ä¸¦çµ¦å‡ºæ¯æ®µçš„é‡é»ã€é †åºã€æ¯é …çš„æ™‚é–“é ä¼°èˆ‡å‚™è¨»ã€‚ç¸½æ™‚æ•¸è«‹å¤§è‡´ç¬¦åˆä»»å‹™ä¼°æ™‚ç¸½å’Œã€‚"
            "å›å‚³ä»¥ Markdown æ¢åˆ—ï¼Œæ ¼å¼ï¼š\n"
            "### Morning\n- [ä»»å‹™ID] ä»»å‹™åç¨±ï¼ˆé ä¼° xx åˆ†ï¼‰â€” å‚™è¨»\n..."
            "### Afternoon\n...\n### Evening\n...\n"
        )
        items = []
        for _, r in cand.iterrows():
            items.append({
                "id": r["id"], "task": r["task"], "est_min": int(r["est_min"]),
                "priority": r["priority"]
            })
        user_content = json.dumps({"today": today, "tasks": items}, ensure_ascii=False)
        try:
            model = genai.GenerativeModel("gemini-1.5-flash")
            resp = model.generate_content(sys_prompt + "\n\n" + user_content)
            plan_md = resp.text
        except Exception as e:
            plan_md = f"âš ï¸ Gemini å¤±æ•—ï¼š{e}\n\næ”¹ç”¨è¦å‰‡å¼è¦åŠƒã€‚"
    else:
        plan_md = "ğŸ”§ æœªè¨­å®š GEMINI_API_KEYï¼Œä½¿ç”¨è¦å‰‡å¼è¦åŠƒã€‚\n\n"

    # è¦å‰‡å¼ï¼šæŠŠé«˜å„ªå…ˆä»»å‹™å¹³å‡åˆ‡åˆ°ä¸Šåˆ/ä¸‹åˆ/æ™šä¸Š
    buckets = {"morning": [], "afternoon": [], "evening": []}
    total = len(cand)
    for i, (_, r) in enumerate(cand.iterrows()):
        if i % 3 == 0:
            buckets["morning"].append(r)
        elif i % 3 == 1:
            buckets["afternoon"].append(r)
        else:
            buckets["evening"].append(r)

    def sec_md(name, rows):
        if not rows: return f"### {name.title()}\nï¼ˆç„¡ï¼‰\n"
        lines = [f"### {name.title()}"]
        for r in rows:
            lines.append(f"- [{r['id']}] {r['task']}ï¼ˆé ä¼° {int(r['est_min'])} åˆ†ï¼ŒP:{r['priority']}ï¼‰")
        return "\n".join(lines) + "\n"

    rule_md = sec_md("morning", buckets["morning"]) + "\n" + \
              sec_md("afternoon", buckets["afternoon"]) + "\n" + \
              sec_md("evening", buckets["evening"])

    return (plan_md + "\n---\n" + rule_md).strip()

# ä»Šæ—¥å®Œæˆç‡
def today_summary():
    global tasks_df
    today = tznow().date().isoformat()
    planned = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today"))
    ]
    done = planned[planned["status"]=="done"]
    total = len(planned)
    done_n = len(done)
    rate = (done_n/total*100) if total>0 else 0
    return f"ğŸ“… ä»Šæ—¥è¨ˆç•«ä»»å‹™ï¼š{total}ï¼›âœ… å®Œæˆï¼š{done_n}ï¼›ğŸ“ˆ å®Œæˆç‡ï¼š{rate:.1f}%"

# =========================
# çˆ¬èŸ²ï¼šæ“·å–æ–‡å­—æˆ–é€£çµä¸¦å¯åŠ å…¥ä»»å‹™
# =========================
def crawl(url, selector, mode, limit):
    try:
        resp = requests.get(url, timeout=15, headers={"User-Agent":"Mozilla/5.0"})
        resp.raise_for_status()
    except Exception as e:
        return pd.DataFrame(columns=CLIPS_HEADER), f"âš ï¸ è«‹æ±‚å¤±æ•—ï¼š{e}"

    soup = BeautifulSoup(resp.text, "html.parser")
    nodes = soup.select(selector)
    rows = []
    for i, n in enumerate(nodes[:int(limit) if limit else 20]):
        text = n.get_text(strip=True) if mode in ("text","both") else ""
        href = n.get("href") if mode in ("href","both") else ""
        # ç›¸å°é€£çµè™•ç†
        if href and href.startswith("/"):
            from urllib.parse import urljoin
            href = urljoin(url, href)
        rows.append({
            "clip_id": str(uuid.uuid4())[:8],
            "url": url,
            "selector": selector,
            "text": text,
            "href": href,
            "created_at": tznow().isoformat(),
            "added_to_task": ""
        })
    df = pd.DataFrame(rows, columns=CLIPS_HEADER)
    return df, f"âœ… æ“·å– {len(df)} ç­†"

def add_clips_as_tasks(clip_ids, default_priority, est_min):
    global clips_df, tasks_df
    if not clip_ids:
        return "âš ï¸ è«‹å…ˆå‹¾é¸è¦åŠ å…¥çš„çˆ¬èŸ²é …ç›®", clips_df, tasks_df
    sel = clips_df[clips_df["clip_id"].isin(clip_ids)]
    _now = tznow().isoformat()
    new_tasks = []
    for _, r in sel.iterrows():
        title = r["text"] or r["href"] or "ï¼ˆæœªå‘½åï¼‰"
        note = f"ä¾†æºï¼š{r['url']}\né¸æ“‡å™¨ï¼š{r['selector']}\né€£çµï¼š{r['href']}"
        new_tasks.append({
            "id": str(uuid.uuid4())[:8],
            "task": title[:120],
            "status": "todo",
            "priority": default_priority or "M",
            "est_min": int(est_min) if est_min else 25,
            "start_time": "",
            "end_time": "",
            "actual_min": 0,
            "pomodoros": 0,
            "due_date": "",
            "labels": "from:crawler",
            "notes": note,
            "created_at": _now,
            "updated_at": _now,
            "completed_at": "",
            "planned_for": ""
        })
    if new_tasks:
        tasks_df = pd.concat([tasks_df, pd.DataFrame(new_tasks)], ignore_index=True)
        # æ¨™è¨˜å·²åŠ å…¥
        clips_df.loc[clips_df["clip_id"].isin(clip_ids), "added_to_task"] = "yes"
        write_df(ws_tasks, tasks_df, TASKS_HEADER)
        write_df(ws_clips, clips_df, CLIPS_HEADER)
        return f"âœ… å·²åŠ å…¥ {len(new_tasks)} é …ç‚ºä»»å‹™", clips_df, tasks_df
    return "âš ï¸ ç„¡å¯åŠ å…¥é …ç›®", clips_df, tasks_df

# =========================
# Gradio ä»‹é¢
# =========================

def _refresh():
    global tasks_df, logs_df, clips_df
    tasks_df, logs_df, clips_df = refresh_all()
    return tasks_df, logs_df, clips_df, list_task_choices(), today_summary()


with gr.Blocks(title="å¾…è¾¦æ¸…å–®ï¼‹ç•ªèŒ„é˜ï¼‹AI è¨ˆç•«ï¼ˆSheet/Gradio/çˆ¬èŸ²ï¼‰") as demo:
    gr.Markdown("# âœ… å¾…è¾¦æ¸…å–®èˆ‡ç•ªèŒ„é˜ï¼ˆGoogle Sheetï¼‹Gradioï¼‹Crawlerï¼‹AI è¨ˆç•«ï¼‰")
    with gr.Row():
        btn_refresh = gr.Button("ğŸ”„ é‡æ–°æ•´ç†ï¼ˆSheet â†’ Appï¼‰")
        out_summary = gr.Markdown(today_summary())

    with gr.Tab("Tasks"):
        with gr.Row():
            with gr.Column(scale=2):
                task = gr.Textbox(label="ä»»å‹™åç¨±", placeholder="å¯« HW3 å ±å‘Š / ä¿®æ­£ SQL / â€¦")
                priority = gr.Dropdown(["H","M","L"], value="M", label="å„ªå…ˆç´š")
                est_min = gr.Number(value=25, label="é ä¼°æ™‚é–“ï¼ˆåˆ†é˜ï¼‰", precision=0)
                due_date = gr.Textbox(label="åˆ°æœŸæ—¥ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰")
                labels = gr.Textbox(label="æ¨™ç±¤ï¼ˆé€—è™Ÿåˆ†éš”ï¼Œå¯ç©ºç™½ï¼‰")
                notes = gr.Textbox(label="å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
                planned_for = gr.Dropdown(["","today","tomorrow"], value="", label="è¦åŠƒæ­¸å±¬")
                btn_add = gr.Button("â• æ–°å¢ä»»å‹™")
                msg_add = gr.Markdown()
            with gr.Column(scale=3):
                grid_tasks = gr.Dataframe(value=tasks_df, label="ä»»å‹™æ¸…å–®ï¼ˆç›´æ¥å¾ Sheet ä¾†ï¼‰", interactive=False)

        with gr.Row():
            task_choice = gr.Dropdown(choices=list_task_choices(), label="é¸å–ä»»å‹™ï¼ˆç”¨æ–¼æ›´æ–°ï¼‰")
            new_status = gr.Dropdown(["todo","in-progress","done"], value="in-progress", label="æ›´æ–°ç‹€æ…‹")
            btn_update = gr.Button("âœï¸ æ›´æ–°ç‹€æ…‹")
            btn_done = gr.Button("âœ… ç›´æ¥æ¨™è¨˜å®Œæˆ")
            msg_update = gr.Markdown()

    with gr.Tab("Pomodoro"):
        with gr.Row():
            sel_task = gr.Dropdown(choices=list_task_choices(), label="é¸æ“‡ä»»å‹™")
            cycles = gr.Number(value=1, precision=0, label="ç•ªèŒ„æ•¸ï¼ˆåƒ…ä½œç´€éŒ„ï¼‰")
        with gr.Row():
            btn_start_work = gr.Button("â–¶ï¸ é–‹å§‹å·¥ä½œ")
            note_work = gr.Textbox(label="å·¥ä½œå‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_work = gr.Button("â¹ï¸ çµæŸå·¥ä½œä¸¦è¨˜éŒ„")
        with gr.Row():
            btn_start_break = gr.Button("ğŸµ é–‹å§‹ä¼‘æ¯")
            note_break = gr.Textbox(label="ä¼‘æ¯å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_break = gr.Button("â¹ï¸ çµæŸä¼‘æ¯ä¸¦è¨˜éŒ„")
        msg_pomo = gr.Markdown()
        grid_logs = gr.Dataframe(value=logs_df, label="ç•ªèŒ„é˜ç´€éŒ„", interactive=False)

    with gr.Tab("AI Plan"):
        gr.Markdown("æŠŠ**ä»Šå¤©çš„ä»»å‹™**æ’æˆ **morning / afternoon / evening** ä¸‰æ®µè¡Œå‹•è¨ˆç•«ã€‚è‹¥æœªè¨­ GEMINI_API_KEYï¼Œæœƒç”¨è¦å‰‡å¼ã€‚")
        btn_plan = gr.Button("ğŸ§  ç”¢ç”Ÿä»Šæ—¥è¨ˆç•«")
        out_plan = gr.Markdown()

    with gr.Tab("Crawler"):
        url = gr.Textbox(label="ç›®æ¨™ URL", placeholder="https://example.com")
        selector = gr.Textbox(label="CSS Selector", placeholder="a.news-item / h2.title / div.card a")
        mode = gr.Radio(["text","href","both"], value="text", label="æ“·å–å…§å®¹")
        limit = gr.Number(value=20, precision=0, label="æœ€å¤šæ“·å–å¹¾ç­†")
        btn_crawl = gr.Button("ğŸ•·ï¸ é–‹å§‹æ“·å–")
        msg_crawl = gr.Markdown()
        grid_clips = gr.Dataframe(value=clips_df, label="æ“·å–çµæœï¼ˆæœƒåŒæ­¥å¯«å…¥ Sheetï¼‰", interactive=True)
        clip_ids = gr.Textbox(label="è¦åŠ å…¥ä»»å‹™çš„ clip_idï¼ˆå¤šå€‹ä»¥é€—è™Ÿåˆ†éš”ï¼‰")
        default_priority = gr.Dropdown(["H","M","L"], value="L", label="æ–°å¢ä»»å‹™å„ªå…ˆç´š")
        clip_est = gr.Number(value=25, precision=0, label="æ–°å¢ä»»å‹™é ä¼°åˆ†é˜")
        btn_add_clips = gr.Button("â• å°‡å‹¾é¸çš„æ“·å–é …ç›®åŠ å…¥ç‚ºä»»å‹™")
        msg_add_clips = gr.Markdown()

    with gr.Tab("Summary"):
        btn_summary = gr.Button("ğŸ“Š é‡æ–°è¨ˆç®—ä»Šæ—¥å®Œæˆç‡")
        out_summary2 = gr.Markdown()

    # === ç¶å®šå‹•ä½œ ===
    btn_refresh.click(_refresh, outputs=[grid_tasks, grid_logs, grid_clips, task_choice, out_summary])

    btn_add.click(
        add_task,
        inputs=[task, priority, est_min, due_date, labels, notes, planned_for],
        outputs=[msg_add, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, out_summary]
    )

    btn_update.click(
        update_task_status,
        inputs=[task_choice, new_status],
        outputs=[msg_update, grid_tasks]
    )

    btn_done.click(
        mark_done,
        inputs=[task_choice],
        outputs=[msg_update, grid_tasks]
    )

    btn_start_work.click(
        start_phase, inputs=[sel_task, gr.State("work"), cycles], outputs=[msg_pomo]
    )
    btn_end_work.click(
        end_phase, inputs=[sel_task, note_work], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, out_summary]
    )
    btn_start_break.click(
        start_phase, inputs=[sel_task, gr.State("break"), cycles], outputs=[msg_pomo]
    )
    btn_end_break.click(
        end_phase, inputs=[sel_task, note_break], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, out_summary]
    )

    btn_plan.click(generate_today_plan, outputs=[out_plan])

    def _crawl_and_save(u, s, m, l):
        df, msg = crawl(u, s, m, l)
        # å¯«å…¥ web_clipsï¼ˆè¦†è“‹å¼è¿½åŠ ï¼šåˆä½µèˆŠè³‡æ–™ï¼‰
        global clips_df
        if not df.empty:
            clips_df = pd.concat([clips_df, df], ignore_index=True)
            write_df(ws_clips, clips_df, CLIPS_HEADER)
        return msg, clips_df

    btn_crawl.click(_crawl_and_save, inputs=[url, selector, mode, limit], outputs=[msg_crawl, grid_clips])

    def _add_clips(clip_ids_str, pr, est):
        ids = [c.strip() for c in (clip_ids_str or "").split(",") if c.strip()]
        msg, new_clips, new_tasks = add_clips_as_tasks(ids, pr, est)
        return msg, new_clips, new_tasks

    btn_add_clips.click(
        _add_clips,
        inputs=[clip_ids, default_priority, clip_est],
        outputs=[msg_add_clips, grid_clips, grid_tasks]
    )

    btn_summary.click(today_summary, outputs=[out_summary2])

demo.launch()
```

```python
def ensure_worksheet(sh, title, header):
    try:
        ws = sh.worksheet(title)
    except gspread.WorksheetNotFound:
        ws = sh.add_worksheet(title=title, rows="1000", cols=str(len(header)+5))
        ws.update([header])
    # è‹¥æ²’æœ‰è¡¨é ­å°±è£œä¸Š
    data = ws.get_all_values()
    if not data or (data and data[0] != header):
        ws.clear()
        ws.update([header])
    return ws

TASKS_HEADER = [
    "id","task","status","priority","est_min","start_time","end_time",
    "actual_min","pomodoros","due_date","labels","notes",
    "created_at","updated_at","completed_at","planned_for"
]
LOGS_HEADER = [
    "log_id","task_id","phase","start_ts","end_ts","minutes","cycles","note"
]
CLIPS_HEADER = ["clip_id","url","selector","text","href","created_at","added_to_task"]

ws_tasks = ensure_worksheet(sh, "tasks", TASKS_HEADER)
ws_logs  = ensure_worksheet(sh, "pomodoro_logs", LOGS_HEADER)
ws_clips = ensure_worksheet(sh, "web_clips", CLIPS_HEADER)

def tznow():
    return dt.now(gettz(TIMEZONE))

def read_df(ws, header):
    df = get_as_dataframe(ws, evaluate_formulas=True, header=0)
    if df is None or df.empty:
        return pd.DataFrame(columns=header)
    df = df.fillna("")
    # ä¿è­‰æ¬„ä½é½Šå…¨
    for c in header:
        if c not in df.columns:
            df[c] = ""
    # å‹åˆ¥å¾®èª¿
    if "est_min" in df.columns:
        df["est_min"] = pd.to_numeric(df["est_min"], errors="coerce").fillna(0).astype(int)
    if "actual_min" in df.columns:
        df["actual_min"] = pd.to_numeric(df["actual_min"], errors="coerce").fillna(0).astype(int)
    if "pomodoros" in df.columns:
        df["pomodoros"] = pd.to_numeric(df["pomodoros"], errors="coerce").fillna(0).astype(int)
    return df[header]

def write_df(ws, df, header):
    if df.empty:
        ws.clear()
        ws.update([header])
        return
    # è½‰å­—ä¸²é¿å… gspread å‹åˆ¥å•é¡Œ
    df_out = df.copy()
    for c in df_out.columns:
        df_out[c] = df_out[c].astype(str)
    ws.clear()
    ws.update([header] + df_out.values.tolist())

def refresh_all():
    return (
        read_df(ws_tasks, TASKS_HEADER).copy(),
        read_df(ws_logs, LOGS_HEADER).copy(),
        read_df(ws_clips, CLIPS_HEADER).copy()
    )

tasks_df, logs_df, clips_df = refresh_all()

def add_task(task, priority, est_min, due_date, labels, notes, planned_for):
    global tasks_df
    _now = tznow().isoformat()
    new = pd.DataFrame([{
        "id": str(uuid.uuid4())[:8],
        "task": task.strip(),
        "status": "todo",
        "priority": priority or "M",
        "est_min": int(est_min) if est_min else 25,
        "start_time": "",
        "end_time": "",
        "actual_min": 0,
        "pomodoros": 0,
        "due_date": due_date or "",
        "labels": labels or "",
        "notes": notes or "",
        "created_at": _now,
        "updated_at": _now,
        "completed_at": "",
        "planned_for": planned_for or ""  # å¯å¡« today / tomorrow / ç©ºç™½
    }])
    tasks_df = pd.concat([tasks_df, new], ignore_index=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… å·²æ–°å¢ä»»å‹™", tasks_df

def update_task_status(task_id, new_status):
    global tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx)==0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    i = idx[0]
    tasks_df.loc[i, "status"] = new_status
    tasks_df.loc[i, "updated_at"] = tznow().isoformat()
    if new_status == "done" and not tasks_df.loc[i, "completed_at"]:
        tasks_df.loc[i, "completed_at"] = tznow().isoformat()
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… ç‹€æ…‹å·²æ›´æ–°", tasks_df

def mark_done(task_id):
    return update_task_status(task_id, "done")

def recalc_task_actuals(task_id):
    """æ ¹æ“š logs_df å›å¯« actual_min èˆ‡ pomodoros"""
    global tasks_df, logs_df
    work_logs = logs_df[(logs_df["task_id"]==task_id) & (logs_df["phase"]=="work")]
    total_min = work_logs["minutes"].astype(float).sum() if not work_logs.empty else 0
    pomos = int(round(total_min / 25.0))
    idx = tasks_df.index[tasks_df["id"]==task_id]
    if len(idx)==0: return
    i = idx[0]
    tasks_df.loc[i,"actual_min"] = int(total_min)
    tasks_df.loc[i,"pomodoros"] = pomos
    tasks_df.loc[i,"updated_at"] = tznow().isoformat()

def list_task_choices():
    global tasks_df
    if tasks_df.empty:
        return []
    # é¡¯ç¤ºï¼š [status] (P:priority) task  â€” id
    def row_label(r):
        return f"[{r['status']}] (P:{r['priority']}) {r['task']} â€” {r['id']}"
    return [(row_label(r), r["id"]) for _, r in tasks_df.iterrows()]

# æˆ‘å€‘æ¡ã€ŒæŒ‰éˆ•é–‹å§‹ / çµæŸã€æ¨¡å¼ï¼ˆé¿å…å¾Œç«¯é˜»å¡ï¼‰ï¼Œæ¯æ¬¡æŒ‰ã€Œé–‹å§‹ã€æœƒå…ˆè¨˜ä½ start_tsï¼Œ
# æŒ‰ã€ŒçµæŸã€æ™‚è¨ˆç®—åˆ†é˜ä¸¦å¯«å…¥ logsï¼Œå†å›å¡«ä»»å‹™ actual_min / pomodorosã€‚

_active_sessions = {}  # { task_id: {"phase": "work"/"break", "start_ts": iso, "cycles": int} }

def start_phase(task_id, phase, cycles):
    if not task_id: return "âš ï¸ è«‹å…ˆé¸æ“‡ä»»å‹™"
    _active_sessions[task_id] = {
        "phase": phase,
        "start_ts": tznow().isoformat(),
        "cycles": int(cycles) if cycles else 1
    }
    return f"â–¶ï¸ å·²é–‹å§‹ï¼š{phase}ï¼ˆtask: {task_id}ï¼‰"

def end_phase(task_id, note):
    global logs_df, tasks_df
    if task_id not in _active_sessions:
        return "âš ï¸ å°šæœªé–‹å§‹ä»»ä½•éšæ®µ"
    sess = _active_sessions.pop(task_id)
    start = pd.to_datetime(sess["start_ts"])
    end = tznow()
    minutes = round((end - start).total_seconds() / 60.0, 2)
    log = pd.DataFrame([{
        "log_id": str(uuid.uuid4())[:8],
        "task_id": task_id,
        "phase": sess["phase"],
        "start_ts": start.isoformat(),
        "end_ts": end.isoformat(),
        "minutes": minutes,
        "cycles": int(sess["cycles"]),
        "note": note or ""
    }])
    logs_df = pd.concat([logs_df, log], ignore_index=True)
    write_df(ws_logs, logs_df, LOGS_HEADER)

    # å›å¡«ä»»å‹™
    if sess["phase"] == "work":
        recalc_task_actuals(task_id)
        write_df(ws_tasks, tasks_df, TASKS_HEADER)

    return f"â¹ï¸ å·²çµæŸï¼š{sess['phase']}ï¼Œç´€éŒ„ {minutes} åˆ†é˜"

# AI è¨ˆç•«ï¼ˆGeminiï¼›ç„¡é‡‘é‘°å‰‡è¦å‰‡å¼ï¼‰
def generate_today_plan():
    global tasks_df
    # ä»¥ã€Œdue_date æ˜¯ä»Šå¤©ã€æˆ–ã€Œplanned_for = todayã€ä¸”ä¸æ˜¯ done çš„ä»»å‹™ç‚ºè¨ˆç•«æ¸…å–®
    today = tznow().date().isoformat()
    cand = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today")) &
        (tasks_df["status"]!="done")
    ].copy()
    if cand.empty:
        return "ğŸ“­ ä»Šå¤©æ²’æœ‰æ¨™è¨˜çš„ä»»å‹™ã€‚è«‹åœ¨ Tasks åˆ†é æŠŠä»»å‹™çš„ due_date è¨­ç‚ºä»Šå¤©æˆ– planned_for è¨­ç‚º todayã€‚"

    # å…ˆä¾ priorityï¼ˆH>M>Lï¼‰+ est_min æ’åº
    pr_order = {"H":0, "M":1, "L":2}
    cand["p_ord"] = cand["priority"].map(pr_order).fillna(3)
    cand = cand.sort_values(["p_ord","est_min"], ascending=[True, True])

    # å˜—è©¦ Gemini
    api_key = os.environ.get("GEMINI_API_KEY","").strip()
    if api_key:
        genai.configure(api_key=api_key)
        sys_prompt = (
            "ä½ æ˜¯ä¸€ä½ä»»å‹™è¦åŠƒåŠ©ç†ã€‚è«‹æŠŠè¼¸å…¥çš„ä»»å‹™ï¼ˆå«ä¼°æ™‚èˆ‡å„ªå…ˆç´šï¼‰æ’æˆä¸‰æ®µï¼šmorningã€afternoonã€eveningï¼Œ"
            "ä¸¦çµ¦å‡ºæ¯æ®µçš„é‡é»ã€é †åºã€æ¯é …çš„æ™‚é–“é ä¼°èˆ‡å‚™è¨»ã€‚ç¸½æ™‚æ•¸è«‹å¤§è‡´ç¬¦åˆä»»å‹™ä¼°æ™‚ç¸½å’Œã€‚"
            "å›å‚³ä»¥ Markdown æ¢åˆ—ï¼Œæ ¼å¼ï¼š\n"
            "### Morning\n- [ä»»å‹™ID] ä»»å‹™åç¨±ï¼ˆé ä¼° xx åˆ†ï¼‰â€” å‚™è¨»\n..."
            "### Afternoon\n...\n### Evening\n...\n"
        )
        items = []
        for _, r in cand.iterrows():
            items.append({
                "id": r["id"], "task": r["task"], "est_min": int(r["est_min"]),
                "priority": r["priority"]
            })
        user_content = json.dumps({"today": today, "tasks": items}, ensure_ascii=False)
        try:
            model = genai.GenerativeModel("gemini-1.5-flash")
            resp = model.generate_content(sys_prompt + "\n\n" + user_content)
            plan_md = resp.text
        except Exception as e:
            plan_md = f"âš ï¸ Gemini å¤±æ•—ï¼š{e}\n\næ”¹ç”¨è¦å‰‡å¼è¦åŠƒã€‚"
    else:
        plan_md = "ğŸ”§ æœªè¨­å®š GEMINI_API_KEYï¼Œä½¿ç”¨è¦å‰‡å¼è¦åŠƒã€‚\n\n"

    # è¦å‰‡å¼ï¼šæŠŠé«˜å„ªå…ˆä»»å‹™å¹³å‡åˆ‡åˆ°ä¸Šåˆ/ä¸‹åˆ/æ™šä¸Š
    buckets = {"morning": [], "afternoon": [], "evening": []}
    total = len(cand)
    for i, (_, r) in enumerate(cand.iterrows()):
        if i % 3 == 0:
            buckets["morning"].append(r)
        elif i % 3 == 1:
            buckets["afternoon"].append(r)
        else:
            buckets["evening"].append(r)

    def sec_md(name, rows):
        if not rows: return f"### {name.title()}\nï¼ˆç„¡ï¼‰\n"
        lines = [f"### {name.title()}"]
        for r in rows:
            lines.append(f"- [{r['id']}] {r['task']}ï¼ˆé ä¼° {int(r['est_min'])} åˆ†ï¼ŒP:{r['priority']}ï¼‰")
        return "\n".join(lines) + "\n"

    rule_md = sec_md("morning", buckets["morning"]) + "\n" + \
              sec_md("afternoon", buckets["afternoon"]) + "\n" + \
              sec_md("evening", buckets["evening"])

    return (plan_md + "\n---\n" + rule_md).strip()

# ä»Šæ—¥å®Œæˆç‡
def today_summary():
    global tasks_df
    today = tznow().date().isoformat()
    planned = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today"))
    ]
    done = planned[planned["status"]=="done"]
    total = len(planned)
    done_n = len(done)
    rate = (done_n/total*100) if total>0 else 0
    return f"ğŸ“… ä»Šæ—¥è¨ˆç•«ä»»å‹™ï¼š{total}ï¼›âœ… å®Œæˆï¼š{done_n}ï¼›ğŸ“ˆ å®Œæˆç‡ï¼š{rate:.1f}%"

# =========================
# çˆ¬èŸ²ï¼šæ“·å–æ–‡å­—æˆ–é€£çµä¸¦å¯åŠ å…¥ä»»å‹™
# =========================
def crawl(url, selector, mode, limit):
    try:
        resp = requests.get(url, timeout=15, headers={"User-Agent":"Mozilla/5.0"})
        resp.raise_for_status()
    except Exception as e:
        return pd.DataFrame(columns=CLIPS_HEADER), f"âš ï¸ è«‹æ±‚å¤±æ•—ï¼š{e}"

    soup = BeautifulSoup(resp.text, "html.parser")
    nodes = soup.select(selector)
    rows = []
    for i, n in enumerate(nodes[:int(limit) if limit else 20]):
        text = n.get_text(strip=True) if mode in ("text","both") else ""
        href = n.get("href") if mode in ("href","both") else ""
        # ç›¸å°é€£çµè™•ç†
        if href and href.startswith("/"):
            from urllib.parse import urljoin
            href = urljoin(url, href)
        rows.append({
            "clip_id": str(uuid.uuid4())[:8],
            "url": url,
            "selector": selector,
            "text": text,
            "href": href,
            "created_at": tznow().isoformat(),
            "added_to_task": ""
        })
    df = pd.DataFrame(rows, columns=CLIPS_HEADER)
    return df, f"âœ… æ“·å– {len(df)} ç­†"

def add_clips_as_tasks(clip_ids, default_priority, est_min):
    global clips_df, tasks_df
    if not clip_ids:
        return "âš ï¸ è«‹å…ˆå‹¾é¸è¦åŠ å…¥çš„çˆ¬èŸ²é …ç›®", clips_df, tasks_df
    sel = clips_df[clips_df["clip_id"].isin(clip_ids)]
    _now = tznow().isoformat()
    new_tasks = []
    for _, r in sel.iterrows():
        title = r["text"] or r["href"] or "ï¼ˆæœªå‘½åï¼‰"
        note = f"ä¾†æºï¼š{r['url']}\né¸æ“‡å™¨ï¼š{r['selector']}\né€£çµï¼š{r['href']}"
        new_tasks.append({
            "id": str(uuid.uuid4())[:8],
            "task": title[:120],
            "status": "todo",
            "priority": default_priority or "M",
            "est_min": int(est_min) if est_min else 25,
            "start_time": "",
            "end_time": "",
            "actual_min": 0,
            "pomodoros": 0,
            "due_date": "",
            "labels": "from:crawler",
            "notes": note,
            "created_at": _now,
            "updated_at": _now,
            "completed_at": "",
            "planned_for": ""
        })
    if new_tasks:
        tasks_df = pd.concat([tasks_df, pd.DataFrame(new_tasks)], ignore_index=True)
        # æ¨™è¨˜å·²åŠ å…¥
        clips_df.loc[clips_df["clip_id"].isin(clip_ids), "added_to_task"] = "yes"
        write_df(ws_tasks, tasks_df, TASKS_HEADER)
        write_df(ws_clips, clips_df, CLIPS_HEADER)
        return f"âœ… å·²åŠ å…¥ {len(new_tasks)} é …ç‚ºä»»å‹™", clips_df, tasks_df
    return "âš ï¸ ç„¡å¯åŠ å…¥é …ç›®", clips_df, tasks_df

def export_tasks_logs(format="csv"):
    global tasks_df, logs_df
    _now = dt.now().strftime("%Y%m%d_%H%M%S")

    if format == "csv":
        tasks_file = f"tasks_{_now}.csv"
        logs_file  = f"logs_{_now}.csv"
        tasks_df.to_csv(tasks_file, index=False)
        logs_df.to_csv(logs_file, index=False)
    else:
        tasks_file = f"tasks_{_now}.json"
        logs_file  = f"logs_{_now}.json"
        tasks_df.to_json(tasks_file, orient="records", force_ascii=False, indent=2)
        logs_df.to_json(logs_file, orient="records", force_ascii=False, indent=2)

    return tasks_file, logs_file

def import_tasks_logs(tasks_file, logs_file, format="csv"):
    global tasks_df, logs_df, ws_tasks, ws_logs
    try:
        # ç¢ºèªè‡³å°‘æœ‰ä¸Šå‚³ä¸€å€‹
        if tasks_file is None and logs_file is None:
            return "âš ï¸ è«‹è‡³å°‘ä¸Šå‚³ä¸€å€‹æª”æ¡ˆï¼ˆä»»å‹™æˆ–ç´€éŒ„ï¼‰"

        # åŒ¯å…¥ä»»å‹™
        if tasks_file is not None:
            if format == "csv":
                df_tasks = pd.read_csv(tasks_file.name)
            elif format == "json":
                with open(tasks_file.name, "r", encoding="utf-8") as f:
                    df_tasks = pd.DataFrame(json.load(f))
            write_df(ws_tasks, df_tasks, TASKS_HEADER)
            tasks_df = df_tasks

        # åŒ¯å…¥ç´€éŒ„
        if logs_file is not None:
            if format == "csv":
                df_logs = pd.read_csv(logs_file.name)
            elif format == "json":
                with open(logs_file.name, "r", encoding="utf-8") as f:
                    df_logs = pd.DataFrame(json.load(f))
            write_df(ws_logs, df_logs, LOGS_HEADER)
            logs_df = df_logs

        return "âœ… åŒ¯å…¥å®Œæˆï¼Œè³‡æ–™å·²å¯«å…¥ Google Sheet ä¸¦åŒæ­¥åˆ°ç•«é¢"
    except Exception as e:
        return f"âš ï¸ åŒ¯å…¥å¤±æ•—ï¼š{e}"

# =========================
# Gradio ä»‹é¢
# =========================

def _refresh():
    global tasks_df, logs_df, clips_df
    tasks_df, logs_df, clips_df = refresh_all()
    choices = list_task_choices()
    summary = today_summary()
    # å›å‚³æ™‚ä¿æŒå›ºå®šçµæ§‹
    return (
        tasks_df,
        logs_df,
        clips_df,
        gr.update(choices=choices),
        gr.update(choices=choices),
        summary
    )


with gr.Blocks(title="å¾…è¾¦æ¸…å–®ï¼‹ç•ªèŒ„é˜ï¼‹AI è¨ˆç•«ï¼ˆSheet/Gradio/çˆ¬èŸ²ï¼‰") as demo:
    gr.Markdown("# âœ… å¾…è¾¦æ¸…å–®èˆ‡ç•ªèŒ„é˜ï¼ˆGoogle Sheetï¼‹Gradioï¼‹Crawlerï¼‹AI è¨ˆç•«ï¼‰")
    with gr.Row():
        btn_refresh = gr.Button("ğŸ”„ é‡æ–°æ•´ç†ï¼ˆSheet â†’ Appï¼‰")
        out_summary = gr.Markdown(today_summary())

    with gr.Tab("Tasks"):
        with gr.Row():
            with gr.Column(scale=2):
                task = gr.Textbox(label="ä»»å‹™åç¨±", placeholder="å¯« HW3 å ±å‘Š / ä¿®æ­£ SQL / â€¦")
                priority = gr.Dropdown(["H","M","L"], value="M", label="å„ªå…ˆç´š")
                est_min = gr.Number(value=25, label="é ä¼°æ™‚é–“ï¼ˆåˆ†é˜ï¼‰", precision=0)
                due_date = gr.Textbox(label="åˆ°æœŸæ—¥ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰")
                labels = gr.Textbox(label="æ¨™ç±¤ï¼ˆé€—è™Ÿåˆ†éš”ï¼Œå¯ç©ºç™½ï¼‰")
                notes = gr.Textbox(label="å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
                planned_for = gr.Dropdown(["","today","tomorrow"], value="", label="è¦åŠƒæ­¸å±¬")
                btn_add = gr.Button("â• æ–°å¢ä»»å‹™")
                msg_add = gr.Markdown()
            with gr.Column(scale=3):
                grid_tasks = gr.Dataframe(value=tasks_df, label="ä»»å‹™æ¸…å–®ï¼ˆç›´æ¥å¾ Sheet ä¾†ï¼‰", interactive=False)

        with gr.Row():
            task_choice = gr.Dropdown(choices=list_task_choices(), label="é¸å–ä»»å‹™ï¼ˆç”¨æ–¼æ›´æ–°ï¼‰")
            new_status = gr.Dropdown(["todo","in-progress","done"], value="in-progress", label="æ›´æ–°ç‹€æ…‹")
            btn_update = gr.Button("âœï¸ æ›´æ–°ç‹€æ…‹")
            btn_done = gr.Button("âœ… ç›´æ¥æ¨™è¨˜å®Œæˆ")
            msg_update = gr.Markdown()

    with gr.Tab("Pomodoro"):
        with gr.Row():
            sel_task = gr.Dropdown(choices=list_task_choices(), label="é¸æ“‡ä»»å‹™")
            cycles = gr.Number(value=1, precision=0, label="ç•ªèŒ„æ•¸ï¼ˆåƒ…ä½œç´€éŒ„ï¼‰")
        with gr.Row():
            btn_start_work = gr.Button("â–¶ï¸ é–‹å§‹å·¥ä½œ")
            note_work = gr.Textbox(label="å·¥ä½œå‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_work = gr.Button("â¹ï¸ çµæŸå·¥ä½œä¸¦è¨˜éŒ„")
        with gr.Row():
            btn_start_break = gr.Button("ğŸµ é–‹å§‹ä¼‘æ¯")
            note_break = gr.Textbox(label="ä¼‘æ¯å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_break = gr.Button("â¹ï¸ çµæŸä¼‘æ¯ä¸¦è¨˜éŒ„")
        msg_pomo = gr.Markdown()
        grid_logs = gr.Dataframe(value=logs_df, label="ç•ªèŒ„é˜ç´€éŒ„", interactive=False)

    with gr.Tab("AI Plan"):
        gr.Markdown("æŠŠ**ä»Šå¤©çš„ä»»å‹™**æ’æˆ **morning / afternoon / evening** ä¸‰æ®µè¡Œå‹•è¨ˆç•«ã€‚è‹¥æœªè¨­ GEMINI_API_KEYï¼Œæœƒç”¨è¦å‰‡å¼ã€‚")
        btn_plan = gr.Button("ğŸ§  ç”¢ç”Ÿä»Šæ—¥è¨ˆç•«")
        out_plan = gr.Markdown()

    with gr.Tab("Crawler"):
        url = gr.Textbox(label="ç›®æ¨™ URL", placeholder="https://example.com")
        selector = gr.Textbox(label="CSS Selector", placeholder="a.news-item / h2.title / div.card a")
        mode = gr.Radio(["text","href","both"], value="text", label="æ“·å–å…§å®¹")
        limit = gr.Number(value=20, precision=0, label="æœ€å¤šæ“·å–å¹¾ç­†")
        btn_crawl = gr.Button("ğŸ•·ï¸ é–‹å§‹æ“·å–")
        msg_crawl = gr.Markdown()
        grid_clips = gr.Dataframe(value=clips_df, label="æ“·å–çµæœï¼ˆæœƒåŒæ­¥å¯«å…¥ Sheetï¼‰", interactive=True)
        clip_ids = gr.Textbox(label="è¦åŠ å…¥ä»»å‹™çš„ clip_idï¼ˆå¤šå€‹ä»¥é€—è™Ÿåˆ†éš”ï¼‰")
        default_priority = gr.Dropdown(["H","M","L"], value="L", label="æ–°å¢ä»»å‹™å„ªå…ˆç´š")
        clip_est = gr.Number(value=25, precision=0, label="æ–°å¢ä»»å‹™é ä¼°åˆ†é˜")
        btn_add_clips = gr.Button("â• å°‡å‹¾é¸çš„æ“·å–é …ç›®åŠ å…¥ç‚ºä»»å‹™")
        msg_add_clips = gr.Markdown()

    with gr.Tab("Summary"):
        btn_summary = gr.Button("ğŸ“Š é‡æ–°è¨ˆç®—ä»Šæ—¥å®Œæˆç‡")
        out_summary2 = gr.Markdown()

    with gr.Tab("åŒ¯å…¥ / åŒ¯å‡º"):
        gr.Markdown("ğŸ“‚ åŒ¯å…¥æˆ–åŒ¯å‡ºä»»å‹™èˆ‡ç•ªèŒ„ç´€éŒ„ (CSV / JSON)")

        # ===== åŒ¯å‡ºå€å¡Š =====
        with gr.Row():
            export_format = gr.Radio(["csv", "json"], value="csv", label="åŒ¯å‡ºæ ¼å¼")
            btn_export = gr.Button("ğŸ“¤ åŒ¯å‡º")
        with gr.Row():
            tasks_out = gr.File(label="ä¸‹è¼‰ä»»å‹™æª”æ¡ˆ")
            logs_out = gr.File(label="ä¸‹è¼‰ç´€éŒ„æª”æ¡ˆ")

        # ===== åŒ¯å…¥å€å¡Š =====
        with gr.Row():
            tasks_file = gr.File(label="é¸æ“‡ä»»å‹™æª”æ¡ˆ")
            logs_file  = gr.File(label="é¸æ“‡ç´€éŒ„æª”æ¡ˆ")
            import_format = gr.Radio(["csv","json"], value="csv", label="åŒ¯å…¥æ ¼å¼")
            btn_import = gr.Button("ğŸ“¥ åŒ¯å…¥")
        out_import = gr.Markdown()

    # ç¶å®šäº‹ä»¶
    btn_export.click(export_tasks_logs, inputs=[export_format], outputs=[tasks_out, logs_out])

    btn_import.click(
        import_tasks_logs,
        inputs=[tasks_file, logs_file, import_format],
        outputs=[out_import]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    # === ç¶å®šå‹•ä½œ ===
    btn_refresh.click(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_add.click(
        add_task,
        inputs=[task, priority, est_min, due_date, labels, notes, planned_for],
        outputs=[msg_add, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )


    btn_update.click(
        update_task_status,
        inputs=[task_choice, new_status],
        outputs=[msg_update, grid_tasks]
    )

    btn_done.click(
        mark_done,
        inputs=[task_choice],
        outputs=[msg_update, grid_tasks]
    )

    btn_start_work.click(
        start_phase, inputs=[sel_task, gr.State("work"), cycles], outputs=[msg_pomo]
    )
    btn_end_work.click(
        end_phase, inputs=[sel_task, note_work], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )
    btn_start_break.click(
        start_phase, inputs=[sel_task, gr.State("break"), cycles], outputs=[msg_pomo]
    )
    btn_end_break.click(
        end_phase, inputs=[sel_task, note_break], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_plan.click(generate_today_plan, outputs=[out_plan])

    def _crawl_and_save(u, s, m, l):
        df, msg = crawl(u, s, m, l)
        # å¯«å…¥ web_clipsï¼ˆè¦†è“‹å¼è¿½åŠ ï¼šåˆä½µèˆŠè³‡æ–™ï¼‰
        global clips_df
        if not df.empty:
            clips_df = pd.concat([clips_df, df], ignore_index=True)
            write_df(ws_clips, clips_df, CLIPS_HEADER)
        return msg, clips_df

    btn_crawl.click(_crawl_and_save, inputs=[url, selector, mode, limit], outputs=[msg_crawl, grid_clips])

    def _add_clips(clip_ids_str, pr, est):
        ids = [c.strip() for c in (clip_ids_str or "").split(",") if c.strip()]
        msg, new_clips, new_tasks = add_clips_as_tasks(ids, pr, est)
        return msg, new_clips, new_tasks

    btn_add_clips.click(
        _add_clips,
        inputs=[clip_ids, default_priority, clip_est],
        outputs=[msg_add_clips, grid_clips, grid_tasks]
    )

    btn_summary.click(today_summary, outputs=[out_summary2])


demo.launch()
```

```python
def ensure_worksheet(sh, title, header):
    try:
        ws = sh.worksheet(title)
    except gspread.WorksheetNotFound:
        ws = sh.add_worksheet(title=title, rows="1000", cols=str(len(header)+5))
        ws.update([header])
    # è‹¥æ²’æœ‰è¡¨é ­å°±è£œä¸Š
    data = ws.get_all_values()
    if not data or (data and data[0] != header):
        ws.clear()
        ws.update([header])
    return ws

TASKS_HEADER = [
    "id","task","status","priority","est_min","start_time","end_time",
    "actual_min","pomodoros","due_date","labels","notes",
    "created_at","updated_at","completed_at","planned_for"
]
LOGS_HEADER = [
    "log_id","task_id","phase","start_ts","end_ts","minutes","cycles","note"
]
CLIPS_HEADER = ["clip_id","url","selector","text","href","created_at","added_to_task"]

ws_tasks = ensure_worksheet(sh, "tasks", TASKS_HEADER)
ws_logs  = ensure_worksheet(sh, "pomodoro_logs", LOGS_HEADER)
ws_clips = ensure_worksheet(sh, "web_clips", CLIPS_HEADER)

def tznow():
    return dt.now(gettz(TIMEZONE))

def read_df(ws, header):
    df = get_as_dataframe(ws, evaluate_formulas=True, header=0)
    if df is None or df.empty:
        return pd.DataFrame(columns=header)
    df = df.fillna("")
    # ä¿è­‰æ¬„ä½é½Šå…¨
    for c in header:
        if c not in df.columns:
            df[c] = ""
    # å‹åˆ¥å¾®èª¿
    if "est_min" in df.columns:
        df["est_min"] = pd.to_numeric(df["est_min"], errors="coerce").fillna(0).astype(int)
    if "actual_min" in df.columns:
        df["actual_min"] = pd.to_numeric(df["actual_min"], errors="coerce").fillna(0).astype(int)
    if "pomodoros" in df.columns:
        df["pomodoros"] = pd.to_numeric(df["pomodoros"], errors="coerce").fillna(0).astype(int)
    return df[header]

def write_df(ws, df, header):
    if df.empty:
        ws.clear()
        ws.update([header])
        return
    # è½‰å­—ä¸²é¿å… gspread å‹åˆ¥å•é¡Œ
    df_out = df.copy()
    for c in df_out.columns:
        df_out[c] = df_out[c].astype(str)
    ws.clear()
    ws.update([header] + df_out.values.tolist())

def refresh_all():
    return (
        read_df(ws_tasks, TASKS_HEADER).copy(),
        read_df(ws_logs, LOGS_HEADER).copy(),
        read_df(ws_clips, CLIPS_HEADER).copy()
    )

tasks_df, logs_df, clips_df = refresh_all()

def add_task(task, priority, est_min, due_date, labels, notes, planned_for):
    global tasks_df
    _now = tznow().isoformat()
    new = pd.DataFrame([{
        "id": str(uuid.uuid4())[:8],
        "task": task.strip(),
        "status": "todo",
        "priority": priority or "M",
        "est_min": int(est_min) if est_min else 25,
        "start_time": "",
        "end_time": "",
        "actual_min": 0,
        "pomodoros": 0,
        "due_date": due_date or "",
        "labels": labels or "",
        "notes": notes or "",
        "created_at": _now,
        "updated_at": _now,
        "completed_at": "",
        "planned_for": planned_for or ""  # å¯å¡« today / tomorrow / ç©ºç™½
    }])
    tasks_df = pd.concat([tasks_df, new], ignore_index=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… å·²æ–°å¢ä»»å‹™", tasks_df

def update_task_status(task_id, new_status):
    global tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx)==0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    i = idx[0]
    tasks_df.loc[i, "status"] = new_status
    tasks_df.loc[i, "updated_at"] = tznow().isoformat()
    if new_status == "done" and not tasks_df.loc[i, "completed_at"]:
        tasks_df.loc[i, "completed_at"] = tznow().isoformat()
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… ç‹€æ…‹å·²æ›´æ–°", tasks_df

def mark_done(task_id):
    return update_task_status(task_id, "done")

def recalc_task_actuals(task_id):
    """æ ¹æ“š logs_df å›å¯« actual_min èˆ‡ pomodoros"""
    global tasks_df, logs_df
    work_logs = logs_df[(logs_df["task_id"]==task_id) & (logs_df["phase"]=="work")]
    total_min = work_logs["minutes"].astype(float).sum() if not work_logs.empty else 0
    pomos = int(round(total_min / 25.0))
    idx = tasks_df.index[tasks_df["id"]==task_id]
    if len(idx)==0: return
    i = idx[0]
    tasks_df.loc[i,"actual_min"] = int(total_min)
    tasks_df.loc[i,"pomodoros"] = pomos
    tasks_df.loc[i,"updated_at"] = tznow().isoformat()

def list_task_choices():
    global tasks_df
    if tasks_df.empty:
        return []
    # é¡¯ç¤ºï¼š [status] (P:priority) task  â€” id
    def row_label(r):
        return f"[{r['status']}] (P:{r['priority']}) {r['task']} â€” {r['id']}"
    return [(row_label(r), r["id"]) for _, r in tasks_df.iterrows()]

# æˆ‘å€‘æ¡ã€ŒæŒ‰éˆ•é–‹å§‹ / çµæŸã€æ¨¡å¼ï¼ˆé¿å…å¾Œç«¯é˜»å¡ï¼‰ï¼Œæ¯æ¬¡æŒ‰ã€Œé–‹å§‹ã€æœƒå…ˆè¨˜ä½ start_tsï¼Œ
# æŒ‰ã€ŒçµæŸã€æ™‚è¨ˆç®—åˆ†é˜ä¸¦å¯«å…¥ logsï¼Œå†å›å¡«ä»»å‹™ actual_min / pomodorosã€‚

_active_sessions = {}  # { task_id: {"phase": "work"/"break", "start_ts": iso, "cycles": int} }

def start_phase(task_id, phase, cycles):
    if not task_id: return "âš ï¸ è«‹å…ˆé¸æ“‡ä»»å‹™"
    _active_sessions[task_id] = {
        "phase": phase,
        "start_ts": tznow().isoformat(),
        "cycles": int(cycles) if cycles else 1
    }
    return f"â–¶ï¸ å·²é–‹å§‹ï¼š{phase}ï¼ˆtask: {task_id}ï¼‰"

def end_phase(task_id, note):
    global logs_df, tasks_df
    if task_id not in _active_sessions:
        return "âš ï¸ å°šæœªé–‹å§‹ä»»ä½•éšæ®µ"
    sess = _active_sessions.pop(task_id)
    start = pd.to_datetime(sess["start_ts"])
    end = tznow()
    minutes = round((end - start).total_seconds() / 60.0, 2)
    log = pd.DataFrame([{
        "log_id": str(uuid.uuid4())[:8],
        "task_id": task_id,
        "phase": sess["phase"],
        "start_ts": start.isoformat(),
        "end_ts": end.isoformat(),
        "minutes": minutes,
        "cycles": int(sess["cycles"]),
        "note": note or ""
    }])
    logs_df = pd.concat([logs_df, log], ignore_index=True)
    write_df(ws_logs, logs_df, LOGS_HEADER)

    # å›å¡«ä»»å‹™
    if sess["phase"] == "work":
        recalc_task_actuals(task_id)
        write_df(ws_tasks, tasks_df, TASKS_HEADER)

    return f"â¹ï¸ å·²çµæŸï¼š{sess['phase']}ï¼Œç´€éŒ„ {minutes} åˆ†é˜"

# AI è¨ˆç•«ï¼ˆGeminiï¼›ç„¡é‡‘é‘°å‰‡è¦å‰‡å¼ï¼‰
def generate_today_plan():
    global tasks_df
    # ä»¥ã€Œdue_date æ˜¯ä»Šå¤©ã€æˆ–ã€Œplanned_for = todayã€ä¸”ä¸æ˜¯ done çš„ä»»å‹™ç‚ºè¨ˆç•«æ¸…å–®
    today = tznow().date().isoformat()
    cand = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today")) &
        (tasks_df["status"]!="done")
    ].copy()
    if cand.empty:
        return "ğŸ“­ ä»Šå¤©æ²’æœ‰æ¨™è¨˜çš„ä»»å‹™ã€‚è«‹åœ¨ Tasks åˆ†é æŠŠä»»å‹™çš„ due_date è¨­ç‚ºä»Šå¤©æˆ– planned_for è¨­ç‚º todayã€‚"

    # å…ˆä¾ priorityï¼ˆH>M>Lï¼‰+ est_min æ’åº
    pr_order = {"H":0, "M":1, "L":2}
    cand["p_ord"] = cand["priority"].map(pr_order).fillna(3)
    cand = cand.sort_values(["p_ord","est_min"], ascending=[True, True])

    # å˜—è©¦ Gemini
    api_key = os.environ.get("GEMINI_API_KEY","").strip()
    if api_key:
        genai.configure(api_key=api_key)
        sys_prompt = (
            "ä½ æ˜¯ä¸€ä½ä»»å‹™è¦åŠƒåŠ©ç†ã€‚è«‹æŠŠè¼¸å…¥çš„ä»»å‹™ï¼ˆå«ä¼°æ™‚èˆ‡å„ªå…ˆç´šï¼‰æ’æˆä¸‰æ®µï¼šmorningã€afternoonã€eveningï¼Œ"
            "ä¸¦çµ¦å‡ºæ¯æ®µçš„é‡é»ã€é †åºã€æ¯é …çš„æ™‚é–“é ä¼°èˆ‡å‚™è¨»ã€‚ç¸½æ™‚æ•¸è«‹å¤§è‡´ç¬¦åˆä»»å‹™ä¼°æ™‚ç¸½å’Œã€‚"
            "å›å‚³ä»¥ Markdown æ¢åˆ—ï¼Œæ ¼å¼ï¼š\n"
            "### Morning\n- [ä»»å‹™ID] ä»»å‹™åç¨±ï¼ˆé ä¼° xx åˆ†ï¼‰â€” å‚™è¨»\n..."
            "### Afternoon\n...\n### Evening\n...\n"
        )
        items = []
        for _, r in cand.iterrows():
            items.append({
                "id": r["id"], "task": r["task"], "est_min": int(r["est_min"]),
                "priority": r["priority"]
            })
        user_content = json.dumps({"today": today, "tasks": items}, ensure_ascii=False)
        try:
            model = genai.GenerativeModel("gemini-1.5-flash")
            resp = model.generate_content(sys_prompt + "\n\n" + user_content)
            plan_md = resp.text
        except Exception as e:
            plan_md = f"âš ï¸ Gemini å¤±æ•—ï¼š{e}\n\næ”¹ç”¨è¦å‰‡å¼è¦åŠƒã€‚"
    else:
        plan_md = "ğŸ”§ æœªè¨­å®š GEMINI_API_KEYï¼Œä½¿ç”¨è¦å‰‡å¼è¦åŠƒã€‚\n\n"

    # è¦å‰‡å¼ï¼šæŠŠé«˜å„ªå…ˆä»»å‹™å¹³å‡åˆ‡åˆ°ä¸Šåˆ/ä¸‹åˆ/æ™šä¸Š
    buckets = {"morning": [], "afternoon": [], "evening": []}
    total = len(cand)
    for i, (_, r) in enumerate(cand.iterrows()):
        if i % 3 == 0:
            buckets["morning"].append(r)
        elif i % 3 == 1:
            buckets["afternoon"].append(r)
        else:
            buckets["evening"].append(r)

    def sec_md(name, rows):
        if not rows: return f"### {name.title()}\nï¼ˆç„¡ï¼‰\n"
        lines = [f"### {name.title()}"]
        for r in rows:
            lines.append(f"- [{r['id']}] {r['task']}ï¼ˆé ä¼° {int(r['est_min'])} åˆ†ï¼ŒP:{r['priority']}ï¼‰")
        return "\n".join(lines) + "\n"

    rule_md = sec_md("morning", buckets["morning"]) + "\n" + \
              sec_md("afternoon", buckets["afternoon"]) + "\n" + \
              sec_md("evening", buckets["evening"])

    return (plan_md + "\n---\n" + rule_md).strip()

# ä»Šæ—¥å®Œæˆç‡
def today_summary():
    global tasks_df
    today = tznow().date().isoformat()
    planned = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today"))
    ]
    done = planned[planned["status"]=="done"]
    total = len(planned)
    done_n = len(done)
    rate = (done_n/total*100) if total>0 else 0
    return f"ğŸ“… ä»Šæ—¥è¨ˆç•«ä»»å‹™ï¼š{total}ï¼›âœ… å®Œæˆï¼š{done_n}ï¼›ğŸ“ˆ å®Œæˆç‡ï¼š{rate:.1f}%"

# =========================
# çˆ¬èŸ²ï¼šæ“·å–æ–‡å­—æˆ–é€£çµä¸¦å¯åŠ å…¥ä»»å‹™
# =========================
def crawl(url, selector, mode, limit):
    try:
        resp = requests.get(url, timeout=15, headers={"User-Agent":"Mozilla/5.0"})
        resp.raise_for_status()
    except Exception as e:
        return pd.DataFrame(columns=CLIPS_HEADER), f"âš ï¸ è«‹æ±‚å¤±æ•—ï¼š{e}"

    soup = BeautifulSoup(resp.text, "html.parser")
    nodes = soup.select(selector)
    rows = []
    for i, n in enumerate(nodes[:int(limit) if limit else 20]):
        text = n.get_text(strip=True) if mode in ("text","both") else ""
        href = n.get("href") if mode in ("href","both") else ""
        # ç›¸å°é€£çµè™•ç†
        if href and href.startswith("/"):
            from urllib.parse import urljoin
            href = urljoin(url, href)
        rows.append({
            "clip_id": str(uuid.uuid4())[:8],
            "url": url,
            "selector": selector,
            "text": text,
            "href": href,
            "created_at": tznow().isoformat(),
            "added_to_task": ""
        })
    df = pd.DataFrame(rows, columns=CLIPS_HEADER)
    return df, f"âœ… æ“·å– {len(df)} ç­†"

def add_clips_as_tasks(clip_ids, default_priority, est_min):
    global clips_df, tasks_df
    if not clip_ids:
        return "âš ï¸ è«‹å…ˆå‹¾é¸è¦åŠ å…¥çš„çˆ¬èŸ²é …ç›®", clips_df, tasks_df
    sel = clips_df[clips_df["clip_id"].isin(clip_ids)]
    _now = tznow().isoformat()
    new_tasks = []
    for _, r in sel.iterrows():
        title = r["text"] or r["href"] or "ï¼ˆæœªå‘½åï¼‰"
        note = f"ä¾†æºï¼š{r['url']}\né¸æ“‡å™¨ï¼š{r['selector']}\né€£çµï¼š{r['href']}"
        new_tasks.append({
            "id": str(uuid.uuid4())[:8],
            "task": title[:120],
            "status": "todo",
            "priority": default_priority or "M",
            "est_min": int(est_min) if est_min else 25,
            "start_time": "",
            "end_time": "",
            "actual_min": 0,
            "pomodoros": 0,
            "due_date": "",
            "labels": "from:crawler",
            "notes": note,
            "created_at": _now,
            "updated_at": _now,
            "completed_at": "",
            "planned_for": ""
        })
    if new_tasks:
        tasks_df = pd.concat([tasks_df, pd.DataFrame(new_tasks)], ignore_index=True)
        # æ¨™è¨˜å·²åŠ å…¥
        clips_df.loc[clips_df["clip_id"].isin(clip_ids), "added_to_task"] = "yes"
        write_df(ws_tasks, tasks_df, TASKS_HEADER)
        write_df(ws_clips, clips_df, CLIPS_HEADER)
        return f"âœ… å·²åŠ å…¥ {len(new_tasks)} é …ç‚ºä»»å‹™", clips_df, tasks_df
    return "âš ï¸ ç„¡å¯åŠ å…¥é …ç›®", clips_df, tasks_df

def export_tasks_logs(format="csv"):
    global tasks_df, logs_df
    _now = dt.now().strftime("%Y%m%d_%H%M%S")

    if format == "csv":
        tasks_file = f"tasks_{_now}.csv"
        logs_file  = f"logs_{_now}.csv"
        tasks_df.to_csv(tasks_file, index=False)
        logs_df.to_csv(logs_file, index=False)
    else:
        tasks_file = f"tasks_{_now}.json"
        logs_file  = f"logs_{_now}.json"
        tasks_df.to_json(tasks_file, orient="records", force_ascii=False, indent=2)
        logs_df.to_json(logs_file, orient="records", force_ascii=False, indent=2)

    return tasks_file, logs_file

def import_tasks_logs(tasks_file, logs_file, format="csv"):
    global tasks_df, logs_df, ws_tasks, ws_logs
    try:
        # ç¢ºèªè‡³å°‘æœ‰ä¸Šå‚³ä¸€å€‹
        if tasks_file is None and logs_file is None:
            return "âš ï¸ è«‹è‡³å°‘ä¸Šå‚³ä¸€å€‹æª”æ¡ˆï¼ˆä»»å‹™æˆ–ç´€éŒ„ï¼‰"

        # åŒ¯å…¥ä»»å‹™
        if tasks_file is not None:
            if format == "csv":
                df_tasks = pd.read_csv(tasks_file.name)
            elif format == "json":
                with open(tasks_file.name, "r", encoding="utf-8") as f:
                    df_tasks = pd.DataFrame(json.load(f))
            write_df(ws_tasks, df_tasks, TASKS_HEADER)
            tasks_df = df_tasks

        # åŒ¯å…¥ç´€éŒ„
        if logs_file is not None:
            if format == "csv":
                df_logs = pd.read_csv(logs_file.name)
            elif format == "json":
                with open(logs_file.name, "r", encoding="utf-8") as f:
                    df_logs = pd.DataFrame(json.load(f))
            write_df(ws_logs, df_logs, LOGS_HEADER)
            logs_df = df_logs

        return "âœ… åŒ¯å…¥å®Œæˆï¼Œè³‡æ–™å·²å¯«å…¥ Google Sheet ä¸¦åŒæ­¥åˆ°ç•«é¢"
    except Exception as e:
        return f"âš ï¸ åŒ¯å…¥å¤±æ•—ï¼š{e}"

def delete_task(task_id):
    global tasks_df
    if not task_id:
        return "âš ï¸ è«‹å…ˆé¸æ“‡è¦åˆªé™¤çš„ä»»å‹™", tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx) == 0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    tasks_df = tasks_df.drop(idx).reset_index(drop=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return f"ğŸ—‘ï¸ å·²åˆªé™¤ä»»å‹™ï¼š{task_id}", tasks_df

# =========================
# Gradio ä»‹é¢
# =========================

def _refresh():
    global tasks_df, logs_df, clips_df
    tasks_df, logs_df, clips_df = refresh_all()
    choices = list_task_choices()
    summary = today_summary()
    # å›å‚³æ™‚ä¿æŒå›ºå®šçµæ§‹
    return (
        tasks_df,
        logs_df,
        clips_df,
        gr.update(choices=choices),
        gr.update(choices=choices),
        summary
    )


with gr.Blocks(title="å¾…è¾¦æ¸…å–®ï¼‹ç•ªèŒ„é˜ï¼‹AI è¨ˆç•«ï¼ˆSheet/Gradio/çˆ¬èŸ²ï¼‰") as demo:
    gr.Markdown("# âœ… å¾…è¾¦æ¸…å–®èˆ‡ç•ªèŒ„é˜ï¼ˆGoogle Sheetï¼‹Gradioï¼‹Crawlerï¼‹AI è¨ˆç•«ï¼‰")
    with gr.Row():
        btn_refresh = gr.Button("ğŸ”„ é‡æ–°æ•´ç†ï¼ˆSheet â†’ Appï¼‰")
        out_summary = gr.Markdown(today_summary())

    with gr.Tab("Tasks"):
        with gr.Row():
            with gr.Column(scale=2):
                task = gr.Textbox(label="ä»»å‹™åç¨±", placeholder="å¯« HW3 å ±å‘Š / ä¿®æ­£ SQL / â€¦")
                priority = gr.Dropdown(["H","M","L"], value="M", label="å„ªå…ˆç´š")
                est_min = gr.Number(value=25, label="é ä¼°æ™‚é–“ï¼ˆåˆ†é˜ï¼‰", precision=0)
                due_date = gr.Textbox(label="åˆ°æœŸæ—¥ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰")
                labels = gr.Textbox(label="æ¨™ç±¤ï¼ˆé€—è™Ÿåˆ†éš”ï¼Œå¯ç©ºç™½ï¼‰")
                notes = gr.Textbox(label="å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
                planned_for = gr.Dropdown(["","today","tomorrow"], value="", label="è¦åŠƒæ­¸å±¬")
                btn_add = gr.Button("â• æ–°å¢ä»»å‹™")
                msg_add = gr.Markdown()
            with gr.Column(scale=3):
                grid_tasks = gr.Dataframe(value=tasks_df, label="ä»»å‹™æ¸…å–®ï¼ˆç›´æ¥å¾ Sheet ä¾†ï¼‰", interactive=False)

        with gr.Row():
            task_choice = gr.Dropdown(choices=list_task_choices(), label="é¸å–ä»»å‹™ï¼ˆç”¨æ–¼æ›´æ–°ï¼‰")
            new_status = gr.Dropdown(["todo","in-progress","done"], value="in-progress", label="æ›´æ–°ç‹€æ…‹")

        with gr.Row():
            btn_update = gr.Button("âœï¸ æ›´æ–°ç‹€æ…‹")
            btn_done = gr.Button("âœ… ç›´æ¥æ¨™è¨˜å®Œæˆ")
            btn_delete = gr.Button("ğŸ—‘ï¸ åˆªé™¤ä»»å‹™")

        msg_update = gr.Markdown()
        msg_delete = gr.Markdown()

    with gr.Tab("Pomodoro"):
        with gr.Row():
            sel_task = gr.Dropdown(choices=list_task_choices(), label="é¸æ“‡ä»»å‹™")
            cycles = gr.Number(value=1, precision=0, label="ç•ªèŒ„æ•¸ï¼ˆåƒ…ä½œç´€éŒ„ï¼‰")
        with gr.Row():
            btn_start_work = gr.Button("â–¶ï¸ é–‹å§‹å·¥ä½œ")
            note_work = gr.Textbox(label="å·¥ä½œå‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_work = gr.Button("â¹ï¸ çµæŸå·¥ä½œä¸¦è¨˜éŒ„")
        with gr.Row():
            btn_start_break = gr.Button("ğŸµ é–‹å§‹ä¼‘æ¯")
            note_break = gr.Textbox(label="ä¼‘æ¯å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_break = gr.Button("â¹ï¸ çµæŸä¼‘æ¯ä¸¦è¨˜éŒ„")
        msg_pomo = gr.Markdown()
        grid_logs = gr.Dataframe(value=logs_df, label="ç•ªèŒ„é˜ç´€éŒ„", interactive=False)

    with gr.Tab("AI Plan"):
        gr.Markdown("æŠŠ**ä»Šå¤©çš„ä»»å‹™**æ’æˆ **morning / afternoon / evening** ä¸‰æ®µè¡Œå‹•è¨ˆç•«ã€‚è‹¥æœªè¨­ GEMINI_API_KEYï¼Œæœƒç”¨è¦å‰‡å¼ã€‚")
        btn_plan = gr.Button("ğŸ§  ç”¢ç”Ÿä»Šæ—¥è¨ˆç•«")
        out_plan = gr.Markdown()

    with gr.Tab("Crawler"):
        url = gr.Textbox(label="ç›®æ¨™ URL", placeholder="https://example.com")
        selector = gr.Textbox(label="CSS Selector", placeholder="a.news-item / h2.title / div.card a")
        mode = gr.Radio(["text","href","both"], value="text", label="æ“·å–å…§å®¹")
        limit = gr.Number(value=20, precision=0, label="æœ€å¤šæ“·å–å¹¾ç­†")
        btn_crawl = gr.Button("ğŸ•·ï¸ é–‹å§‹æ“·å–")
        msg_crawl = gr.Markdown()
        grid_clips = gr.Dataframe(value=clips_df, label="æ“·å–çµæœï¼ˆæœƒåŒæ­¥å¯«å…¥ Sheetï¼‰", interactive=True)
        clip_ids = gr.Textbox(label="è¦åŠ å…¥ä»»å‹™çš„ clip_idï¼ˆå¤šå€‹ä»¥é€—è™Ÿåˆ†éš”ï¼‰")
        default_priority = gr.Dropdown(["H","M","L"], value="L", label="æ–°å¢ä»»å‹™å„ªå…ˆç´š")
        clip_est = gr.Number(value=25, precision=0, label="æ–°å¢ä»»å‹™é ä¼°åˆ†é˜")
        btn_add_clips = gr.Button("â• å°‡å‹¾é¸çš„æ“·å–é …ç›®åŠ å…¥ç‚ºä»»å‹™")
        msg_add_clips = gr.Markdown()

    with gr.Tab("Summary"):
        btn_summary = gr.Button("ğŸ“Š é‡æ–°è¨ˆç®—ä»Šæ—¥å®Œæˆç‡")
        out_summary2 = gr.Markdown()

    with gr.Tab("åŒ¯å…¥ / åŒ¯å‡º"):
        gr.Markdown("ğŸ“‚ åŒ¯å…¥æˆ–åŒ¯å‡ºä»»å‹™èˆ‡ç•ªèŒ„ç´€éŒ„ (CSV / JSON)")

        # ===== åŒ¯å‡ºå€å¡Š =====
        with gr.Row():
            export_format = gr.Radio(["csv", "json"], value="csv", label="åŒ¯å‡ºæ ¼å¼")
            btn_export = gr.Button("ğŸ“¤ åŒ¯å‡º")
        with gr.Row():
            tasks_out = gr.File(label="ä¸‹è¼‰ä»»å‹™æª”æ¡ˆ")
            logs_out = gr.File(label="ä¸‹è¼‰ç´€éŒ„æª”æ¡ˆ")

        # ===== åŒ¯å…¥å€å¡Š =====
        with gr.Row():
            tasks_file = gr.File(label="é¸æ“‡ä»»å‹™æª”æ¡ˆ")
            logs_file  = gr.File(label="é¸æ“‡ç´€éŒ„æª”æ¡ˆ")
            import_format = gr.Radio(["csv","json"], value="csv", label="åŒ¯å…¥æ ¼å¼")
            btn_import = gr.Button("ğŸ“¥ åŒ¯å…¥")
        out_import = gr.Markdown()

    # ç¶å®šäº‹ä»¶
    btn_export.click(export_tasks_logs, inputs=[export_format], outputs=[tasks_out, logs_out])

    btn_import.click(
        import_tasks_logs,
        inputs=[tasks_file, logs_file, import_format],
        outputs=[out_import]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    # === ç¶å®šå‹•ä½œ ===
    btn_refresh.click(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_add.click(
        add_task,
        inputs=[task, priority, est_min, due_date, labels, notes, planned_for],
        outputs=[msg_add, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_delete.click(
        delete_task,
        inputs=[task_choice],
        outputs=[msg_delete, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_update.click(
        update_task_status,
        inputs=[task_choice, new_status],
        outputs=[msg_update, grid_tasks]
    )

    btn_done.click(
        mark_done,
        inputs=[task_choice],
        outputs=[msg_update, grid_tasks]
    )

    btn_start_work.click(
        start_phase, inputs=[sel_task, gr.State("work"), cycles], outputs=[msg_pomo]
    )
    btn_end_work.click(
        end_phase, inputs=[sel_task, note_work], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )
    btn_start_break.click(
        start_phase, inputs=[sel_task, gr.State("break"), cycles], outputs=[msg_pomo]
    )
    btn_end_break.click(
        end_phase, inputs=[sel_task, note_break], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_plan.click(generate_today_plan, outputs=[out_plan])

    def _crawl_and_save(u, s, m, l):
        df, msg = crawl(u, s, m, l)
        # å¯«å…¥ web_clipsï¼ˆè¦†è“‹å¼è¿½åŠ ï¼šåˆä½µèˆŠè³‡æ–™ï¼‰
        global clips_df
        if not df.empty:
            clips_df = pd.concat([clips_df, df], ignore_index=True)
            write_df(ws_clips, clips_df, CLIPS_HEADER)
        return msg, clips_df

    btn_crawl.click(_crawl_and_save, inputs=[url, selector, mode, limit], outputs=[msg_crawl, grid_clips])

    def _add_clips(clip_ids_str, pr, est):
        ids = [c.strip() for c in (clip_ids_str or "").split(",") if c.strip()]
        msg, new_clips, new_tasks = add_clips_as_tasks(ids, pr, est)
        return msg, new_clips, new_tasks

    btn_add_clips.click(
        _add_clips,
        inputs=[clip_ids, default_priority, clip_est],
        outputs=[msg_add_clips, grid_clips, grid_tasks]
    )

    btn_summary.click(today_summary, outputs=[out_summary2])


demo.launch()
```

```python
def ensure_worksheet(sh, title, header):
    try:
        ws = sh.worksheet(title)
    except gspread.WorksheetNotFound:
        ws = sh.add_worksheet(title=title, rows="1000", cols=str(len(header)+5))
        ws.update([header])
    # è‹¥æ²’æœ‰è¡¨é ­å°±è£œä¸Š
    data = ws.get_all_values()
    if not data or (data and data[0] != header):
        ws.clear()
        ws.update([header])
    return ws

TASKS_HEADER = [
    "id","task","status","priority","est_min","start_time","end_time",
    "actual_min","pomodoros","due_date","labels","notes",
    "created_at","updated_at","completed_at","planned_for"
]
LOGS_HEADER = [
    "log_id","task_id","phase","start_ts","end_ts","minutes","cycles","note"
]
CLIPS_HEADER = ["clip_id","url","selector","text","href","created_at","added_to_task"]

ws_tasks = ensure_worksheet(sh, "tasks", TASKS_HEADER)
ws_logs  = ensure_worksheet(sh, "pomodoro_logs", LOGS_HEADER)
ws_clips = ensure_worksheet(sh, "web_clips", CLIPS_HEADER)

def tznow():
    return dt.now(gettz(TIMEZONE))

def read_df(ws, header):
    df = get_as_dataframe(ws, evaluate_formulas=True, header=0)
    if df is None or df.empty:
        return pd.DataFrame(columns=header)
    df = df.fillna("")
    # ä¿è­‰æ¬„ä½é½Šå…¨
    for c in header:
        if c not in df.columns:
            df[c] = ""
    # å‹åˆ¥å¾®èª¿
    if "est_min" in df.columns:
        df["est_min"] = pd.to_numeric(df["est_min"], errors="coerce").fillna(0).astype(int)
    if "actual_min" in df.columns:
        df["actual_min"] = pd.to_numeric(df["actual_min"], errors="coerce").fillna(0).astype(int)
    if "pomodoros" in df.columns:
        df["pomodoros"] = pd.to_numeric(df["pomodoros"], errors="coerce").fillna(0).astype(int)
    return df[header]

def write_df(ws, df, header):
    if df.empty:
        ws.clear()
        ws.update([header])
        return
    # è½‰å­—ä¸²é¿å… gspread å‹åˆ¥å•é¡Œ
    df_out = df.copy()
    for c in df_out.columns:
        df_out[c] = df_out[c].astype(str)
    ws.clear()
    ws.update([header] + df_out.values.tolist())

def refresh_all():
    return (
        read_df(ws_tasks, TASKS_HEADER).copy(),
        read_df(ws_logs, LOGS_HEADER).copy(),
        read_df(ws_clips, CLIPS_HEADER).copy()
    )

tasks_df, logs_df, clips_df = refresh_all()

def add_task(task, priority, est_min, due_date, labels, notes, planned_for):
    global tasks_df
    _now = tznow().isoformat()
    new = pd.DataFrame([{
        "id": str(uuid.uuid4())[:8],
        "task": task.strip(),
        "status": "todo",
        "priority": priority or "M",
        "est_min": int(est_min) if est_min else 25,
        "start_time": "",
        "end_time": "",
        "actual_min": 0,
        "pomodoros": 0,
        "due_date": due_date or "",
        "labels": labels or "",
        "notes": notes or "",
        "created_at": _now,
        "updated_at": _now,
        "completed_at": "",
        "planned_for": planned_for or ""  # å¯å¡« today / tomorrow / ç©ºç™½
    }])
    tasks_df = pd.concat([tasks_df, new], ignore_index=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… å·²æ–°å¢ä»»å‹™", tasks_df

def update_task_status(task_id, new_status):
    global tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx)==0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    i = idx[0]
    tasks_df.loc[i, "status"] = new_status
    tasks_df.loc[i, "updated_at"] = tznow().isoformat()
    if new_status == "done" and not tasks_df.loc[i, "completed_at"]:
        tasks_df.loc[i, "completed_at"] = tznow().isoformat()
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return "âœ… ç‹€æ…‹å·²æ›´æ–°", tasks_df

def mark_done(task_id):
    return update_task_status(task_id, "done")

def recalc_task_actuals(task_id):
    """æ ¹æ“š logs_df å›å¯« actual_min èˆ‡ pomodoros"""
    global tasks_df, logs_df
    work_logs = logs_df[(logs_df["task_id"]==task_id) & (logs_df["phase"]=="work")]
    total_min = work_logs["minutes"].astype(float).sum() if not work_logs.empty else 0
    pomos = int(round(total_min / 25.0))
    idx = tasks_df.index[tasks_df["id"]==task_id]
    if len(idx)==0: return
    i = idx[0]
    tasks_df.loc[i,"actual_min"] = int(total_min)
    tasks_df.loc[i,"pomodoros"] = pomos
    tasks_df.loc[i,"updated_at"] = tznow().isoformat()

def list_task_choices():
    global tasks_df
    if tasks_df.empty:
        return []
    # é¡¯ç¤ºï¼š [status] (P:priority) task  â€” id
    def row_label(r):
        return f"[{r['status']}] (P:{r['priority']}) {r['task']} â€” {r['id']}"
    return [(row_label(r), r["id"]) for _, r in tasks_df.iterrows()]

# æˆ‘å€‘æ¡ã€ŒæŒ‰éˆ•é–‹å§‹ / çµæŸã€æ¨¡å¼ï¼ˆé¿å…å¾Œç«¯é˜»å¡ï¼‰ï¼Œæ¯æ¬¡æŒ‰ã€Œé–‹å§‹ã€æœƒå…ˆè¨˜ä½ start_tsï¼Œ
# æŒ‰ã€ŒçµæŸã€æ™‚è¨ˆç®—åˆ†é˜ä¸¦å¯«å…¥ logsï¼Œå†å›å¡«ä»»å‹™ actual_min / pomodorosã€‚

_active_sessions = {}  # { task_id: {"phase": "work"/"break", "start_ts": iso, "cycles": int} }

def start_phase(task_id, phase, cycles):
    if not task_id: return "âš ï¸ è«‹å…ˆé¸æ“‡ä»»å‹™"
    _active_sessions[task_id] = {
        "phase": phase,
        "start_ts": tznow().isoformat(),
        "cycles": int(cycles) if cycles else 1
    }
    return f"â–¶ï¸ å·²é–‹å§‹ï¼š{phase}ï¼ˆtask: {task_id}ï¼‰"

def end_phase(task_id, note):
    global logs_df, tasks_df
    if task_id not in _active_sessions:
        return "âš ï¸ å°šæœªé–‹å§‹ä»»ä½•éšæ®µ"
    sess = _active_sessions.pop(task_id)
    start = pd.to_datetime(sess["start_ts"])
    end = tznow()
    minutes = round((end - start).total_seconds() / 60.0, 2)
    log = pd.DataFrame([{
        "log_id": str(uuid.uuid4())[:8],
        "task_id": task_id,
        "phase": sess["phase"],
        "start_ts": start.isoformat(),
        "end_ts": end.isoformat(),
        "minutes": minutes,
        "cycles": int(sess["cycles"]),
        "note": note or ""
    }])
    logs_df = pd.concat([logs_df, log], ignore_index=True)
    write_df(ws_logs, logs_df, LOGS_HEADER)

    # å›å¡«ä»»å‹™
    if sess["phase"] == "work":
        recalc_task_actuals(task_id)
        write_df(ws_tasks, tasks_df, TASKS_HEADER)

    return f"â¹ï¸ å·²çµæŸï¼š{sess['phase']}ï¼Œç´€éŒ„ {minutes} åˆ†é˜"

# AI è¨ˆç•«ï¼ˆGeminiï¼›ç„¡é‡‘é‘°å‰‡è¦å‰‡å¼ï¼‰
def generate_today_plan():
    global tasks_df
    # ä»¥ã€Œdue_date æ˜¯ä»Šå¤©ã€æˆ–ã€Œplanned_for = todayã€ä¸”ä¸æ˜¯ done çš„ä»»å‹™ç‚ºè¨ˆç•«æ¸…å–®
    today = tznow().date().isoformat()
    cand = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today")) &
        (tasks_df["status"]!="done")
    ].copy()
    if cand.empty:
        return "ğŸ“­ ä»Šå¤©æ²’æœ‰æ¨™è¨˜çš„ä»»å‹™ã€‚è«‹åœ¨ Tasks åˆ†é æŠŠä»»å‹™çš„ due_date è¨­ç‚ºä»Šå¤©æˆ– planned_for è¨­ç‚º todayã€‚"

    # å…ˆä¾ priorityï¼ˆH>M>Lï¼‰+ est_min æ’åº
    pr_order = {"H":0, "M":1, "L":2}
    cand["p_ord"] = cand["priority"].map(pr_order).fillna(3)
    cand = cand.sort_values(["p_ord","est_min"], ascending=[True, True])

    # å˜—è©¦ Gemini
    api_key = os.environ.get("GEMINI_API_KEY","").strip()
    if api_key:
        genai.configure(api_key=api_key)
        sys_prompt = (
            "ä½ æ˜¯ä¸€ä½ä»»å‹™è¦åŠƒåŠ©ç†ã€‚è«‹æŠŠè¼¸å…¥çš„ä»»å‹™ï¼ˆå«ä¼°æ™‚èˆ‡å„ªå…ˆç´šï¼‰æ’æˆä¸‰æ®µï¼šmorningã€afternoonã€eveningï¼Œ"
            "ä¸¦çµ¦å‡ºæ¯æ®µçš„é‡é»ã€é †åºã€æ¯é …çš„æ™‚é–“é ä¼°èˆ‡å‚™è¨»ã€‚ç¸½æ™‚æ•¸è«‹å¤§è‡´ç¬¦åˆä»»å‹™ä¼°æ™‚ç¸½å’Œã€‚"
            "å›å‚³ä»¥ Markdown æ¢åˆ—ï¼Œæ ¼å¼ï¼š\n"
            "### Morning\n- [ä»»å‹™ID] ä»»å‹™åç¨±ï¼ˆé ä¼° xx åˆ†ï¼‰â€” å‚™è¨»\n..."
            "### Afternoon\n...\n### Evening\n...\n"
        )
        items = []
        for _, r in cand.iterrows():
            items.append({
                "id": r["id"], "task": r["task"], "est_min": int(r["est_min"]),
                "priority": r["priority"]
            })
        user_content = json.dumps({"today": today, "tasks": items}, ensure_ascii=False)
        try:
            model = genai.GenerativeModel("gemini-1.5-flash")
            resp = model.generate_content(sys_prompt + "\n\n" + user_content)
            plan_md = resp.text
        except Exception as e:
            plan_md = f"âš ï¸ Gemini å¤±æ•—ï¼š{e}\n\næ”¹ç”¨è¦å‰‡å¼è¦åŠƒã€‚"
    else:
        plan_md = "ğŸ”§ æœªè¨­å®š GEMINI_API_KEYï¼Œä½¿ç”¨è¦å‰‡å¼è¦åŠƒã€‚\n\n"

    # è¦å‰‡å¼ï¼šæŠŠé«˜å„ªå…ˆä»»å‹™å¹³å‡åˆ‡åˆ°ä¸Šåˆ/ä¸‹åˆ/æ™šä¸Š
    buckets = {"morning": [], "afternoon": [], "evening": []}
    total = len(cand)
    for i, (_, r) in enumerate(cand.iterrows()):
        if i % 3 == 0:
            buckets["morning"].append(r)
        elif i % 3 == 1:
            buckets["afternoon"].append(r)
        else:
            buckets["evening"].append(r)

    def sec_md(name, rows):
        if not rows: return f"### {name.title()}\nï¼ˆç„¡ï¼‰\n"
        lines = [f"### {name.title()}"]
        for r in rows:
            lines.append(f"- [{r['id']}] {r['task']}ï¼ˆé ä¼° {int(r['est_min'])} åˆ†ï¼ŒP:{r['priority']}ï¼‰")
        return "\n".join(lines) + "\n"

    rule_md = sec_md("morning", buckets["morning"]) + "\n" + \
              sec_md("afternoon", buckets["afternoon"]) + "\n" + \
              sec_md("evening", buckets["evening"])

    return (plan_md + "\n---\n" + rule_md).strip()

# ä»Šæ—¥å®Œæˆç‡
def today_summary():
    global tasks_df
    today = tznow().date().isoformat()
    planned = tasks_df[
        ((tasks_df["due_date"]==today) | (tasks_df["planned_for"].str.lower()=="today"))
    ]
    done = planned[planned["status"]=="done"]
    total = len(planned)
    done_n = len(done)
    rate = (done_n/total*100) if total>0 else 0
    return f"ğŸ“… ä»Šæ—¥è¨ˆç•«ä»»å‹™ï¼š{total}ï¼›âœ… å®Œæˆï¼š{done_n}ï¼›ğŸ“ˆ å®Œæˆç‡ï¼š{rate:.1f}%"

# =========================
# çˆ¬èŸ²ï¼šæ“·å–æ–‡å­—æˆ–é€£çµä¸¦å¯åŠ å…¥ä»»å‹™
# =========================
def crawl(url, selector, mode, limit):
    try:
        resp = requests.get(url, timeout=15, headers={"User-Agent":"Mozilla/5.0"})
        resp.raise_for_status()
    except Exception as e:
        return pd.DataFrame(columns=CLIPS_HEADER), f"âš ï¸ è«‹æ±‚å¤±æ•—ï¼š{e}"

    soup = BeautifulSoup(resp.text, "html.parser")
    nodes = soup.select(selector)
    rows = []
    for i, n in enumerate(nodes[:int(limit) if limit else 20]):
        text = n.get_text(strip=True) if mode in ("text","both") else ""
        href = n.get("href") if mode in ("href","both") else ""
        # ç›¸å°é€£çµè™•ç†
        if href and href.startswith("/"):
            from urllib.parse import urljoin
            href = urljoin(url, href)
        rows.append({
            "clip_id": str(uuid.uuid4())[:8],
            "url": url,
            "selector": selector,
            "text": text,
            "href": href,
            "created_at": tznow().isoformat(),
            "added_to_task": ""
        })
    df = pd.DataFrame(rows, columns=CLIPS_HEADER)
    return df, f"âœ… æ“·å– {len(df)} ç­†"

def add_clips_as_tasks(clip_ids, default_priority, est_min):
    global clips_df, tasks_df
    if not clip_ids:
        return "âš ï¸ è«‹å…ˆå‹¾é¸è¦åŠ å…¥çš„çˆ¬èŸ²é …ç›®", clips_df, tasks_df
    sel = clips_df[clips_df["clip_id"].isin(clip_ids)]
    _now = tznow().isoformat()
    new_tasks = []
    for _, r in sel.iterrows():
        title = r["text"] or r["href"] or "ï¼ˆæœªå‘½åï¼‰"
        note = f"ä¾†æºï¼š{r['url']}\né¸æ“‡å™¨ï¼š{r['selector']}\né€£çµï¼š{r['href']}"
        new_tasks.append({
            "id": str(uuid.uuid4())[:8],
            "task": title[:120],
            "status": "todo",
            "priority": default_priority or "M",
            "est_min": int(est_min) if est_min else 25,
            "start_time": "",
            "end_time": "",
            "actual_min": 0,
            "pomodoros": 0,
            "due_date": "",
            "labels": "from:crawler",
            "notes": note,
            "created_at": _now,
            "updated_at": _now,
            "completed_at": "",
            "planned_for": ""
        })
    if new_tasks:
        tasks_df = pd.concat([tasks_df, pd.DataFrame(new_tasks)], ignore_index=True)
        # æ¨™è¨˜å·²åŠ å…¥
        clips_df.loc[clips_df["clip_id"].isin(clip_ids), "added_to_task"] = "yes"
        write_df(ws_tasks, tasks_df, TASKS_HEADER)
        write_df(ws_clips, clips_df, CLIPS_HEADER)
        return f"âœ… å·²åŠ å…¥ {len(new_tasks)} é …ç‚ºä»»å‹™", clips_df, tasks_df
    return "âš ï¸ ç„¡å¯åŠ å…¥é …ç›®", clips_df, tasks_df

def export_tasks_logs(format="csv"):
    global tasks_df, logs_df
    _now = dt.now().strftime("%Y%m%d_%H%M%S")

    if format == "csv":
        tasks_file = f"tasks_{_now}.csv"
        logs_file  = f"logs_{_now}.csv"
        tasks_df.to_csv(tasks_file, index=False)
        logs_df.to_csv(logs_file, index=False)
    else:
        tasks_file = f"tasks_{_now}.json"
        logs_file  = f"logs_{_now}.json"
        tasks_df.to_json(tasks_file, orient="records", force_ascii=False, indent=2)
        logs_df.to_json(logs_file, orient="records", force_ascii=False, indent=2)

    return tasks_file, logs_file

def import_tasks_logs(tasks_file, logs_file, format="csv"):
    global tasks_df, logs_df, ws_tasks, ws_logs
    try:
        # ç¢ºèªè‡³å°‘æœ‰ä¸Šå‚³ä¸€å€‹
        if tasks_file is None and logs_file is None:
            return "âš ï¸ è«‹è‡³å°‘ä¸Šå‚³ä¸€å€‹æª”æ¡ˆï¼ˆä»»å‹™æˆ–ç´€éŒ„ï¼‰"

        # åŒ¯å…¥ä»»å‹™
        if tasks_file is not None:
            if format == "csv":
                df_tasks = pd.read_csv(tasks_file.name)
            elif format == "json":
                with open(tasks_file.name, "r", encoding="utf-8") as f:
                    df_tasks = pd.DataFrame(json.load(f))
            write_df(ws_tasks, df_tasks, TASKS_HEADER)
            tasks_df = df_tasks

        # åŒ¯å…¥ç´€éŒ„
        if logs_file is not None:
            if format == "csv":
                df_logs = pd.read_csv(logs_file.name)
            elif format == "json":
                with open(logs_file.name, "r", encoding="utf-8") as f:
                    df_logs = pd.DataFrame(json.load(f))
            write_df(ws_logs, df_logs, LOGS_HEADER)
            logs_df = df_logs

        return "âœ… åŒ¯å…¥å®Œæˆï¼Œè³‡æ–™å·²å¯«å…¥ Google Sheet ä¸¦åŒæ­¥åˆ°ç•«é¢"
    except Exception as e:
        return f"âš ï¸ åŒ¯å…¥å¤±æ•—ï¼š{e}"

def delete_task(task_id):
    global tasks_df
    if not task_id:
        return "âš ï¸ è«‹å…ˆé¸æ“‡è¦åˆªé™¤çš„ä»»å‹™", tasks_df
    idx = tasks_df.index[tasks_df["id"] == task_id]
    if len(idx) == 0:
        return "âš ï¸ æ‰¾ä¸åˆ°ä»»å‹™", tasks_df
    tasks_df = tasks_df.drop(idx).reset_index(drop=True)
    write_df(ws_tasks, tasks_df, TASKS_HEADER)
    return f"ğŸ—‘ï¸ å·²åˆªé™¤ä»»å‹™ï¼š{task_id}", tasks_df

def query_tasks(keyword, status_filter, priority_filter, start_due, end_due):
    global tasks_df
    df = tasks_df.copy()

    # ç‹€æ…‹ç¯©é¸
    if status_filter:
        df = df[df["status"].fillna("").str.lower().isin([s.lower() for s in status_filter])]

    # å„ªå…ˆç´šç¯©é¸ï¼ˆè‹¥æœ‰å¡«ï¼‰
    if priority_filter and priority_filter.strip():
        df = df[df["priority"].fillna("").str.upper() == priority_filter.upper()]

    # é—œéµå­—æœå°‹ï¼ˆtask + notesï¼‰
    if keyword and keyword.strip():
        kw = keyword.lower()
        df = df[
            df["task"].str.lower().str.contains(kw, na=False)
            | df["notes"].str.lower().str.contains(kw, na=False)
        ]

    # åˆ°æœŸæ—¥ç¯„åœï¼ˆdue_dateï¼‰
    df_dates = pd.to_datetime(df["due_date"], errors="coerce")  # ç„¡æ³•è§£æçš„è®Š NaT
    if start_due:
        start_dt = pd.to_datetime(start_due)
        df = df[df_dates >= start_dt]
    if end_due:
        end_dt = pd.to_datetime(end_due)
        df = df[df_dates <= end_dt]

    if df.empty:
        return "ğŸ“­ æ²’æœ‰ç¬¦åˆæ¢ä»¶çš„ä»»å‹™", pd.DataFrame(columns=tasks_df.columns)

    df = df.sort_values("due_date", ascending=True)
    return f"âœ… æ‰¾åˆ° {len(df)} ç­†ä»»å‹™", df


def export_query_results(df):
    """æŠŠæŸ¥è©¢çµæœåŒ¯å‡ºæˆ CSV"""
    if df is None or len(df) == 0:
        return None
    _now = dt.now().strftime("%Y%m%d_%H%M%S")
    filename = f"query_results_{_now}.csv"
    df.to_csv(filename, index=False)
    return filename

# =========================
# Gradio ä»‹é¢
# =========================

def _refresh():
    global tasks_df, logs_df, clips_df
    tasks_df, logs_df, clips_df = refresh_all()
    choices = list_task_choices()
    summary = today_summary()
    # å›å‚³æ™‚ä¿æŒå›ºå®šçµæ§‹
    return (
        tasks_df,
        logs_df,
        clips_df,
        gr.update(choices=choices),
        gr.update(choices=choices),
        summary
    )


with gr.Blocks(title="å¾…è¾¦æ¸…å–®ï¼‹ç•ªèŒ„é˜ï¼‹AI è¨ˆç•«ï¼ˆSheet/Gradio/çˆ¬èŸ²ï¼‰") as demo:
    gr.Markdown("# âœ… å¾…è¾¦æ¸…å–®èˆ‡ç•ªèŒ„é˜ï¼ˆGoogle Sheetï¼‹Gradioï¼‹Crawlerï¼‹AI è¨ˆç•«ï¼‰")
    with gr.Row():
        btn_refresh = gr.Button("ğŸ”„ é‡æ–°æ•´ç†ï¼ˆSheet â†’ Appï¼‰")
        out_summary = gr.Markdown(today_summary())

    with gr.Tab("Tasks"):
        with gr.Row():
            with gr.Column(scale=2):
                task = gr.Textbox(label="ä»»å‹™åç¨±", placeholder="å¯« HW3 å ±å‘Š / ä¿®æ­£ SQL / â€¦")
                priority = gr.Dropdown(["H","M","L"], value="M", label="å„ªå…ˆç´š")
                est_min = gr.Number(value=25, label="é ä¼°æ™‚é–“ï¼ˆåˆ†é˜ï¼‰", precision=0)
                due_date = gr.Textbox(label="åˆ°æœŸæ—¥ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰")
                labels = gr.Textbox(label="æ¨™ç±¤ï¼ˆé€—è™Ÿåˆ†éš”ï¼Œå¯ç©ºç™½ï¼‰")
                notes = gr.Textbox(label="å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
                planned_for = gr.Dropdown(["","today","tomorrow"], value="", label="è¦åŠƒæ­¸å±¬")
                btn_add = gr.Button("â• æ–°å¢ä»»å‹™")
                msg_add = gr.Markdown()
            with gr.Column(scale=3):
                grid_tasks = gr.Dataframe(value=tasks_df, label="ä»»å‹™æ¸…å–®ï¼ˆç›´æ¥å¾ Sheet ä¾†ï¼‰", interactive=False)

        with gr.Row():
            task_choice = gr.Dropdown(choices=list_task_choices(), label="é¸å–ä»»å‹™ï¼ˆç”¨æ–¼æ›´æ–°ï¼‰")
            new_status = gr.Dropdown(["todo","in-progress","done"], value="in-progress", label="æ›´æ–°ç‹€æ…‹")

        with gr.Row():
            btn_update = gr.Button("âœï¸ æ›´æ–°ç‹€æ…‹")
            btn_done = gr.Button("âœ… ç›´æ¥æ¨™è¨˜å®Œæˆ")
            btn_delete = gr.Button("ğŸ—‘ï¸ åˆªé™¤ä»»å‹™")

        msg_update = gr.Markdown()
        msg_delete = gr.Markdown()

    with gr.Tab("Pomodoro"):
        with gr.Row():
            sel_task = gr.Dropdown(choices=list_task_choices(), label="é¸æ“‡ä»»å‹™")
            cycles = gr.Number(value=1, precision=0, label="ç•ªèŒ„æ•¸ï¼ˆåƒ…ä½œç´€éŒ„ï¼‰")
        with gr.Row():
            btn_start_work = gr.Button("â–¶ï¸ é–‹å§‹å·¥ä½œ")
            note_work = gr.Textbox(label="å·¥ä½œå‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_work = gr.Button("â¹ï¸ çµæŸå·¥ä½œä¸¦è¨˜éŒ„")
        with gr.Row():
            btn_start_break = gr.Button("ğŸµ é–‹å§‹ä¼‘æ¯")
            note_break = gr.Textbox(label="ä¼‘æ¯å‚™è¨»ï¼ˆå¯ç©ºç™½ï¼‰")
            btn_end_break = gr.Button("â¹ï¸ çµæŸä¼‘æ¯ä¸¦è¨˜éŒ„")
        msg_pomo = gr.Markdown()
        grid_logs = gr.Dataframe(value=logs_df, label="ç•ªèŒ„é˜ç´€éŒ„", interactive=False)

    with gr.Tab("AI Plan"):
        gr.Markdown("æŠŠ**ä»Šå¤©çš„ä»»å‹™**æ’æˆ **morning / afternoon / evening** ä¸‰æ®µè¡Œå‹•è¨ˆç•«ã€‚è‹¥æœªè¨­ GEMINI_API_KEYï¼Œæœƒç”¨è¦å‰‡å¼ã€‚")
        btn_plan = gr.Button("ğŸ§  ç”¢ç”Ÿä»Šæ—¥è¨ˆç•«")
        out_plan = gr.Markdown()

    with gr.Tab("Crawler"):
        url = gr.Textbox(label="ç›®æ¨™ URL", placeholder="https://example.com")
        selector = gr.Textbox(label="CSS Selector", placeholder="a.news-item / h2.title / div.card a")
        mode = gr.Radio(["text","href","both"], value="text", label="æ“·å–å…§å®¹")
        limit = gr.Number(value=20, precision=0, label="æœ€å¤šæ“·å–å¹¾ç­†")
        btn_crawl = gr.Button("ğŸ•·ï¸ é–‹å§‹æ“·å–")
        msg_crawl = gr.Markdown()
        grid_clips = gr.Dataframe(value=clips_df, label="æ“·å–çµæœï¼ˆæœƒåŒæ­¥å¯«å…¥ Sheetï¼‰", interactive=True)
        clip_ids = gr.Textbox(label="è¦åŠ å…¥ä»»å‹™çš„ clip_idï¼ˆå¤šå€‹ä»¥é€—è™Ÿåˆ†éš”ï¼‰")
        default_priority = gr.Dropdown(["H","M","L"], value="L", label="æ–°å¢ä»»å‹™å„ªå…ˆç´š")
        clip_est = gr.Number(value=25, precision=0, label="æ–°å¢ä»»å‹™é ä¼°åˆ†é˜")
        btn_add_clips = gr.Button("â• å°‡å‹¾é¸çš„æ“·å–é …ç›®åŠ å…¥ç‚ºä»»å‹™")
        msg_add_clips = gr.Markdown()

    with gr.Tab("Summary"):
        btn_summary = gr.Button("ğŸ“Š é‡æ–°è¨ˆç®—ä»Šæ—¥å®Œæˆç‡")
        out_summary2 = gr.Markdown()

    with gr.Tab("åŒ¯å…¥ / åŒ¯å‡º"):
        gr.Markdown("ğŸ“‚ åŒ¯å…¥æˆ–åŒ¯å‡ºä»»å‹™èˆ‡ç•ªèŒ„ç´€éŒ„ (CSV / JSON)")

        # ===== åŒ¯å‡ºå€å¡Š =====
        with gr.Row():
            export_format = gr.Radio(["csv", "json"], value="csv", label="åŒ¯å‡ºæ ¼å¼")
            btn_export = gr.Button("ğŸ“¤ åŒ¯å‡º")
        with gr.Row():
            tasks_out = gr.File(label="ä¸‹è¼‰ä»»å‹™æª”æ¡ˆ")
            logs_out = gr.File(label="ä¸‹è¼‰ç´€éŒ„æª”æ¡ˆ")

        # ===== åŒ¯å…¥å€å¡Š =====
        with gr.Row():
            tasks_file = gr.File(label="é¸æ“‡ä»»å‹™æª”æ¡ˆ")
            logs_file  = gr.File(label="é¸æ“‡ç´€éŒ„æª”æ¡ˆ")
            import_format = gr.Radio(["csv","json"], value="csv", label="åŒ¯å…¥æ ¼å¼")
            btn_import = gr.Button("ğŸ“¥ åŒ¯å…¥")
        out_import = gr.Markdown()

    with gr.Tab("æŸ¥è©¢ä»»å‹™"):
        gr.Markdown("ğŸ” **ä»»å‹™æŸ¥è©¢** â€” å¯ä¾ç‹€æ…‹ã€å„ªå…ˆç´šã€é—œéµå­—èˆ‡åˆ°æœŸæ—¥ç¯„åœé€²è¡Œç¯©é¸")

        with gr.Row():
            keyword = gr.Textbox(label="é—œéµå­—æœå°‹ï¼ˆä»»å‹™åç¨±æˆ–å‚™è¨»ï¼‰", placeholder="ä¾‹å¦‚ï¼šå ±å‘Šã€meetingã€å®¢æˆ¶â€¦")

        with gr.Row():
            status_filter = gr.CheckboxGroup(
                ["todo", "in-progress", "done"],
                value=[],
                label="é¸æ“‡ä»»å‹™ç‹€æ…‹ï¼ˆå¯è¤‡é¸ï¼Œå¯ç•™ç©ºï¼‰"
            )
            priority_filter = gr.Dropdown(
                ["", "H", "M", "L"],
                value="",
                label="å„ªå…ˆç´šï¼ˆå¯ç•™ç©ºï¼‰"
            )

        with gr.Row():
            start_due = gr.Textbox(label="åˆ°æœŸæ—¥é–‹å§‹ï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰", placeholder="2025-10-01")
            end_due = gr.Textbox(label="åˆ°æœŸæ—¥çµæŸï¼ˆYYYY-MM-DDï¼Œå¯ç©ºç™½ï¼‰", placeholder="2025-10-31")

        with gr.Row():
            btn_query = gr.Button("ğŸ” æŸ¥è©¢")
            btn_export_query = gr.Button("ğŸ“¤ åŒ¯å‡ºçµæœ (CSV)")

        msg_query = gr.Markdown()
        grid_query = gr.Dataframe(label="æŸ¥è©¢çµæœ", interactive=False)
        file_out = gr.File(label="ä¸‹è¼‰æŸ¥è©¢çµæœ")

    # ç¶å®šäº‹ä»¶
    btn_export.click(export_tasks_logs, inputs=[export_format], outputs=[tasks_out, logs_out])

    btn_import.click(
        import_tasks_logs,
        inputs=[tasks_file, logs_file, import_format],
        outputs=[out_import]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    # === ç¶å®šå‹•ä½œ ===
    btn_refresh.click(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_add.click(
        add_task,
        inputs=[task, priority, est_min, due_date, labels, notes, planned_for],
        outputs=[msg_add, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_delete.click(
        delete_task,
        inputs=[task_choice],
        outputs=[msg_delete, grid_tasks]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_update.click(
        update_task_status,
        inputs=[task_choice, new_status],
        outputs=[msg_update, grid_tasks]
    )

    btn_done.click(
        mark_done,
        inputs=[task_choice],
        outputs=[msg_update, grid_tasks]
    )

    btn_start_work.click(
        start_phase, inputs=[sel_task, gr.State("work"), cycles], outputs=[msg_pomo]
    )
    btn_end_work.click(
        end_phase, inputs=[sel_task, note_work], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )
    btn_start_break.click(
        start_phase, inputs=[sel_task, gr.State("break"), cycles], outputs=[msg_pomo]
    )
    btn_end_break.click(
        end_phase, inputs=[sel_task, note_break], outputs=[msg_pomo]
    ).then(
        _refresh,
        outputs=[grid_tasks, grid_logs, grid_clips, task_choice, sel_task, out_summary]
    )

    btn_plan.click(generate_today_plan, outputs=[out_plan])

    def _crawl_and_save(u, s, m, l):
        df, msg = crawl(u, s, m, l)
        # å¯«å…¥ web_clipsï¼ˆè¦†è“‹å¼è¿½åŠ ï¼šåˆä½µèˆŠè³‡æ–™ï¼‰
        global clips_df
        if not df.empty:
            clips_df = pd.concat([clips_df, df], ignore_index=True)
            write_df(ws_clips, clips_df, CLIPS_HEADER)
        return msg, clips_df

    btn_crawl.click(_crawl_and_save, inputs=[url, selector, mode, limit], outputs=[msg_crawl, grid_clips])

    def _add_clips(clip_ids_str, pr, est):
        ids = [c.strip() for c in (clip_ids_str or "").split(",") if c.strip()]
        msg, new_clips, new_tasks = add_clips_as_tasks(ids, pr, est)
        return msg, new_clips, new_tasks

    btn_add_clips.click(
        _add_clips,
        inputs=[clip_ids, default_priority, clip_est],
        outputs=[msg_add_clips, grid_clips, grid_tasks]
    )

    btn_summary.click(today_summary, outputs=[out_summary2])

    btn_query.click(
        query_tasks,
        inputs=[keyword, status_filter, priority_filter, start_due, end_due],
        outputs=[msg_query, grid_query]
    )

    btn_export_query.click(
        export_query_results,
        inputs=[grid_query],
        outputs=[file_out]
    )

demo.launch()
```
